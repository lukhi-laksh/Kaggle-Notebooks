# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Category')
plt.title(f'{plot_no}. Order Distribution Across Product Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='TotalAmount', estimator=sum)
plt.title(f'{plot_no}. Total Revenue Contribution by Category')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Platform', y='TotalAmount', estimator=sum)
plt.title(f'{plot_no}. Platform-wise Revenue Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='City')
plt.title(f'{plot_no}. Order Volume by City')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Category', y='Price')
plt.title(f'{plot_no}. Price Distribution Across Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Platform', y='Rating')
plt.title(f'{plot_no}. Rating Variability by Platform')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Price'], kde=True)
plt.title(f'{plot_no}. Price Distribution of All Products')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['TotalAmount'], kde=True)
plt.title(f'{plot_no}. Distribution of Transaction Values')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Price', y='TotalAmount', hue='Category')
plt.title(f'{plot_no}. Relationship Between Price and Total Transaction Amount')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Brand', y='TotalAmount', estimator=sum)
plt.title(f'{plot_no}. Brand-wise Revenue Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Brand', y='Rating')
plt.title(f'{plot_no}. Average Customer Rating by Brand')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.sort_values('OrderDate'), x='OrderDate', y='TotalAmount')
plt.title(f'{plot_no}. Revenue Trend Over Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Platform', hue='Category')
plt.title(f'{plot_no}. Category Mix Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='City', y='TotalAmount')
plt.title(f'{plot_no}. City-wise Transaction Value Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Quantity', y='TotalAmount')
plt.title(f'{plot_no}. Impact of Quantity on Transaction Amount')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Category', y='Rating')
plt.title(f'{plot_no}. Rating Density Across Product Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['Price','Quantity','TotalAmount','Rating','Reviews']].corr(), annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Features')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Platform', y='Reviews', estimator=sum)
plt.title(f'{plot_no}. Platform-wise Customer Review Volume')
show_fig()
plot_no += 1
```
