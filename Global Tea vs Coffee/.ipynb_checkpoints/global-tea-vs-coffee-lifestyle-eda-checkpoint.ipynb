# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Beverage Consumers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='gender', data=df)
plt.title(f'{plot_no}. Gender-wise Distribution of Consumers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='income_level', y='monthly_spend', data=df)
plt.title(f'{plot_no}. Monthly Spend Variation Across Income Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='drink_preference', data=df)
plt.title(f'{plot_no}. Preference Between Coffee and Tea')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='drink_category', y='cups_per_day', data=df)
plt.title(f'{plot_no}. Daily Consumption by Drink Category')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sugar_level', y='bmi', data=df)
plt.title(f'{plot_no}. Sugar Level vs BMI Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='cups_per_day', y='sleep_hours', data=df)
plt.title(f'{plot_no}. Impact of Drink Consumption on Sleep Hours')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='caffeine_preference', y='heart_rate', data=df)
plt.title(f'{plot_no}. Heart Rate Distribution by Caffeine Preference')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='stress_level', y='cups_per_day', data=df)
plt.title(f'{plot_no}. Stress Level vs Beverage Intake')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='exercise_frequency', y='bmi', data=df)
plt.title(f'{plot_no}. BMI Distribution Based on Exercise Frequency')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='hydration_level', y='heart_rate', data=df)
plt.title(f'{plot_no}. Hydration Level vs Heart Rate')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='drink_temperature', y='taste_score', data=df)
plt.title(f'{plot_no}. Taste Score Variation by Drink Temperature')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='price_per_cup', y='monthly_spend', data=df)
plt.title(f'{plot_no}. Price per Cup vs Monthly Spending Behavior')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='time_of_day', data=df)
plt.title(f'{plot_no}. Preferred Time of Day for Beverage Consumption')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='work_type', y='cups_per_day', data=df)
plt.title(f'{plot_no}. Beverage Consumption Across Work Types')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='aroma_score', y='satisfaction_level', data=df)
plt.title(f'{plot_no}. Aroma Score vs Overall Satisfaction')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='loyalty_brand', y='monthly_spend', data=df)
plt.title(f'{plot_no}. Brand Loyalty Impact on Monthly Spending')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sleep_hours', y='stress_level', data=df)
plt.title(f'{plot_no}. Relationship Between Sleep Hours and Stress Level')
show_fig()
plot_no += 1
```
