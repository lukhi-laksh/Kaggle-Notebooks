## Map grades to numeric values for analysis

```python
grade_map = {'A+':5, 'A':4, 'B+':3, 'B':2, 'C+':1, 'C':0, 'D':-1, 'F':-2}
for col in ['AL401- [T]', 'AL402- [T]', 'AL403- [P]', 'AL404- [P]', 'AL405- [P]', 'AL406- [P]']:
    df[col+'_num'] = df[col].map(grade_map)
grade_map
```

# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['SGPA'], bins=10, kde=True)
plt.title(f'{plot_no}. Distribution of SGPA Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['CGPA'], bins=10, kde=True)
plt.title(f'{plot_no}. Distribution of CGPA Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['AL401- [T]', 'AL402- [T]', 'AL403- [P]', 'AL404- [P]', 'AL405- [P]', 'AL406- [P]']].apply(lambda x: x.map({'A+':5, 'A':4, 'B+':3, 'B':2, 'C+':1, 'C':0, 'D':-1, 'F':-2})), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Grade Correlation Heatmap of All Subjects')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='SGPA', y='CGPA', hue='Branch', style='Status', s=100)
plt.title(f'{plot_no}. Relationship Between SGPA and CGPA by Branch')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Result Description', order=df['Result Description'].value_counts().index)
plt.title(f'{plot_no}. Student Result Description Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['SGPA','CGPA']].corr(), annot=True, cmap='YlGnBu')
plt.title(f'{plot_no}. Correlation Between SGPA and CGPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Semester', hue='Result Description')
plt.title(f'{plot_no}. Semester-wise Result Distribution')
show_fig()
plot_no += 1
```
