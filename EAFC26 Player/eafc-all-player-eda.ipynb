# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['OVR'], bins=15, kde=True)
plt.title(f'{plot_no}. Distribution of Overall Player Ratings (OVR)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='GENDER', y='OVR', data=df)
plt.title(f'{plot_no}. Comparison of Overall Ratings by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='OVR', hue='GENDER', data=df)
plt.title(f'{plot_no}. Relationship Between Age and Overall Rating')
show_fig()
plot_no += 1
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='PAC', y='DRI', hue='Position', data=df)
plt.title(f'{plot_no}. Pace vs Dribbling Across Playing Positions')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='SHO', y='PAS', hue='GENDER', data=df)
plt.title(f'{plot_no}. Shooting vs Passing Skill Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Position', y='PAC', data=df)
plt.title(f'{plot_no}. Pace Distribution by Playing Position')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Position', y='DEF', data=df)
plt.title(f'{plot_no}. Defensive Ability Comparison by Position')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='Nation', data=df, order=df['Nation'].value_counts().head(10).index)
plt.title(f'{plot_no}. Top 10 Nations by Player Representation')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='League', data=df, order=df['League'].value_counts().index)
plt.title(f'{plot_no}. Player Distribution Across Leagues')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='PAC', y='SHO', size='OVR', data=df)
plt.title(f'{plot_no}. Pace vs Shooting with Overall Rating Influence')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Age'], bins=15, kde=True)
plt.title(f'{plot_no}. Age Distribution of Professional Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Preferred foot', y='OVR', data=df)
plt.title(f'{plot_no}. Overall Rating by Preferred Foot')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Skill moves', y='DRI', data=df)
plt.title(f'{plot_no}. Impact of Skill Moves on Dribbling Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Weak foot', y='PAS', data=df)
plt.title(f'{plot_no}. Weak Foot Rating vs Passing Ability')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='GENDER', y='PAC', data=df)
plt.title(f'{plot_no}. Gender-wise Pace Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='GENDER', y='SHO', data=df)
plt.title(f'{plot_no}. Gender-wise Shooting Ability Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='DEF', y='OVR', data=df)
plt.title(f'{plot_no}. Contribution of Defensive Skills to Overall Rating')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='PAS', y='DRI', data=df)
plt.title(f'{plot_no}. Playmaking Relationship Between Passing and Dribbling')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['PAC'], bins=15, kde=True)
plt.title(f'{plot_no}. Distribution of Player Pace Attributes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='PAC', data=df)
plt.title(f'{plot_no}. Effect of Age on Player Pace')
show_fig()
plot_no += 1
```
