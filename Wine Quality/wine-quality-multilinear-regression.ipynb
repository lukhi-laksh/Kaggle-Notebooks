# EDA

```python
def show_fig(): 
    plt.tight_layout() 
    plt.show() 
plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='alcohol', bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Alcohol — central tendency, spread, and outliers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='pH', bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of pH — acidity profile across samples')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='residual sugar', bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Residual Sugar — sweetness skew and tails')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='quality', order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. Count of Wines by Quality Score — class balance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='quality', y='alcohol', order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. Alcohol by Quality — median and dispersion per quality class')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='quality', y='volatile acidity', order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. Volatile Acidity by Quality — lower VA associated with higher quality?')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.stripplot(data=df, x='quality', y='alcohol', jitter=True, alpha=0.5, order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. Alcohol per Quality (jitter) — point-level view of spread and clusters')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(data=df, x='volatile acidity', y='quality', scatter_kws={'s':20, 'alpha':0.5}, lowess=True)
plt.title(f'{plot_no}. Relationship: Volatile Acidity vs Quality — trend and local smoothing')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(data=df, x='citric acid', y='quality', scatter_kws={'s':20, 'alpha':0.6})
plt.title(f'{plot_no}. Relationship: Citric Acid vs Quality — positive/negative association check')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', center=0)
plt.title(f'{plot_no}. Correlation Matrix — feature interrelationships and multi-collinearity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='residual sugar', y='density', alpha=0.6)
plt.title(f'{plot_no}. Residual Sugar vs Density — physical relationship and concentration effects')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(data=df, x='sulphates', y='alcohol', scatter_kws={'s':20, 'alpha':0.6})
plt.title(f'{plot_no}. Sulphates vs Alcohol — association between preservative level and strength')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='chlorides', y='free sulfur dioxide', alpha=0.6)
plt.title(f'{plot_no}. Chlorides vs Free Sulfur Dioxide — saltiness indicator vs preservative')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='total sulfur dioxide', y='free sulfur dioxide', alpha=0.6)
plt.title(f'{plot_no}. Total Sulfur Dioxide vs Free Sulfur Dioxide — partitioning and potential spoilage risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='quality', y='residual sugar', order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. Residual Sugar by Quality — sweetness dispersion across quality scores')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='alcohol', hue=df['quality'].astype(str), common_norm=False, fill=True, alpha=0.4)
plt.title(f'{plot_no}. Alcohol Distribution by Quality (KDE) — comparative density by class')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='quality', y='pH', order=sorted(df['quality'].unique()))
plt.title(f'{plot_no}. pH Distribution by Quality (Violin) — acidity structure across quality groups')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='fixed acidity', bins=30, kde=True)
plt.title(f'{plot_no}. Fixed Acidity Distribution — titratable acidity insights and skewness')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='alcohol', y='volatile acidity', bins=30, pthresh=.05)
plt.title(f'{plot_no}. 2D Density: Alcohol vs Volatile Acidity — joint concentration hotspots')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr_quality = df.corr()['quality'].drop('quality').sort_values()
sns.barplot(x=corr_quality.values, y=corr_quality.index)
plt.title(f'{plot_no}. Feature Correlation with Quality — sorted influence on quality score')
show_fig()
plot_no += 1
```

# MulitiLinear Regression Prediction

```python
x = df.iloc[:, :-1]
x.head()
```

```python
y = df.iloc[:, -1:]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
Lr = LinearRegression()
Lr.fit(x_train, y_train)
```

```python
Lr.score(x_test, y_test) * 100
```
