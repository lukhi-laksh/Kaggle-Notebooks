# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True)
plt.title(f'{plot_no}. Age Distribution Reveals Population Spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Gender', data=df)
plt.title(f'{plot_no}. Gender Distribution Across Dataset')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Occupation', data=df)
plt.title(f'{plot_no}. Occupation Breakdown Highlights Workforce Composition')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Daily_Screen_Time'], kde=True)
plt.title(f'{plot_no}. Daily Screen Time Distribution Indicates Digital Exposure')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Daily_Screen_Time', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Screen Time vs Sleep Hours Shows Sleep Deprivation Trends')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Stress_Level', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Stress Level Impact on Sleep Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Work_Study_Hours', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Work/Study Hours Influence on Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Depression', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Sleep Patterns Among Depressed vs Non-Depressed Individuals')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Anxiety', y='Daily_Screen_Time', data=df)
plt.title(f'{plot_no}. Screen Time Differences Based on Anxiety Presence')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Burnout', data=df)
plt.title(f'{plot_no}. Burnout Prevalence Across Individuals')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Physical_Activity', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Physical Activity vs Stress Level Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Smoking', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Smoking Behavior Impact on Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Alcohol', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Alcohol Consumption Influence on Sleep Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Caffeine_Intake', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Caffeine Intake vs Sleep Hours Shows Sleep Disruption')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Social_Media_Usage', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Social Media Usage Correlation with Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Depression', y='Social_Interaction_Score', data=df)
plt.title(f'{plot_no}. Social Interaction Differences in Depressed Individuals')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Night_Usage', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Night Device Usage Impact on Sleep Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Daily_Screen_Time', data=df)
plt.title(f'{plot_no}. Age vs Screen Time Reveals Behavioral Trends')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Burnout', y='Work_Study_Hours', data=df)
plt.title(f'{plot_no}. Workload Differences Between Burnout Groups')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(20, 12))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title(f'{plot_no}. Correlation Heatmap Showing Feature Relationships')
show_fig()
plot_no += 1
```
