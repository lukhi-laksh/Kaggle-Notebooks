# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['kills'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Player Kills in Matches')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['damageDealt'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Total Damage Dealt by Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['walkDistance'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Walking Distance Across Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['rideDistance'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Riding Distance Across Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='kills', y='winPlacePerc', data=df, alpha=0.6)
plt.title(f'{plot_no}. Relationship Between Number of Kills and Final Match Placement')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='damageDealt', y='winPlacePerc', data=df, alpha=0.6)
plt.title(f'{plot_no}. Impact of Damage Dealt on Winning Percentage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='walkDistance', y='winPlacePerc', data=df, alpha=0.6)
plt.title(f'{plot_no}. Influence of Walking Distance on Match Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='rideDistance', y='winPlacePerc', data=df, alpha=0.6)
plt.title(f'{plot_no}. Relationship Between Ride Distance and Winning Placement')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='weaponsAcquired', y='winPlacePerc', data=df)
plt.title(f'{plot_no}. Effect of Weapons Acquired on Winning Percentage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='heals', y='winPlacePerc', data=df)
plt.title(f'{plot_no}. Healing Usage vs Final Match Placement')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='boosts', y='winPlacePerc', data=df)
plt.title(f'{plot_no}. Boost Usage Impact on Winning Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='rideDistance', y='walkDistance', data=df, alpha=0.6)
plt.title(f'{plot_no}. Comparison Between Walking and Riding Distances')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='kills', y='winPlacePerc', data=df, estimator=np.mean)
plt.title(f'{plot_no}. Average Winning Percentage by Kill Count')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='boosts', y='winPlacePerc', data=df, estimator=np.mean)
plt.title(f'{plot_no}. Average Winning Percentage by Boost Usage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df[['kills','damageDealt','walkDistance','rideDistance','weaponsAcquired','heals','boosts','winPlacePerc']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Matrix of Player Performance Metrics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='kills', y='winPlacePerc', data=df)
plt.title(f'{plot_no}. Distribution of Winning Percentage Across Kill Counts')
show_fig()
plot_no += 1
```
