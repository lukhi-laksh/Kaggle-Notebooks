# Linear Regression

## Import SkLearn Librarys

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Splits x and y

```python
X = df.drop('Solar_Viability_Score', axis=1)
y = df['Solar_Viability_Score']
```

## Remove unnecessary Object Column 

```python
cat_cols = X.select_dtypes(include=['object']).columns
num_cols = X.select_dtypes(exclude=['object']).columns
```

## Creating Preprocessing Transformer

```python
preprocess = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
], remainder='passthrough')
```

## Building Pipeline with Linear Regression

```python
model = Pipeline([
    ('preprocess', preprocess),
    ('regressor', LinearRegression())
])
```

## Splitting Dataset into 80% Training and 20% Testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

## Training the Linear Regression Model

```python
model.fit(X_train, y_train)
```

## Predicting on Test Data

```python
y_pred = model.predict(X_test)
```

## Calculating R2 Score (%)

```python
score = r2_score(y_test, y_pred)*100
score
```

```python
print(f"RÂ² Score: {score:.2f}% â€” This model captures almost all variability in the target. Excellent fit!")
```
