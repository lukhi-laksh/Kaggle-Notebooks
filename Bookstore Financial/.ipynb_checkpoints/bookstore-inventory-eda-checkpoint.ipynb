# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
df['month'] = pd.to_datetime(df['month'])
monthly_qty = df.groupby('month')['qty'].sum().reset_index()
sns.lineplot(data=monthly_qty, x='month', y='qty', marker='o')
plt.title(f'{plot_no}. Monthly total quantity sold â€” trend and seasonality')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
monthly_revenue = df.groupby('month')['extended_retail'].sum().reset_index()
sns.lineplot(data=monthly_revenue, x='month', y='extended_retail', marker='o')
plt.title(f'{plot_no}. Monthly total revenue (extended_retail) â€” growth and dips')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
df['profit'] = df['extended_retail'] - df['extended_cost']
monthly_profit = df.groupby('month')['profit'].sum().reset_index()
sns.lineplot(data=monthly_profit, x='month', y='profit', marker='o')
plt.title(f'{plot_no}. Monthly total profit â€” profitability trend over time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
region_qty = df.groupby('to')['qty'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=region_qty, x='qty', y='to')
plt.title(f'{plot_no}. Total quantity sold by region â€” highest to lowest demand')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
df['margin'] = (df['profit'] / df['extended_retail']).replace([np.inf, -np.inf], np.nan)
sku_margin = df.groupby('sku')['margin'].median().dropna().sort_values(ascending=False).head(15).reset_index()
sns.barplot(data=sku_margin, x='margin', y='sku')
plt.title(f'{plot_no}. Median profit margin by SKU (top 15) â€” which SKUs sustain margins')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_regions = df['to'].value_counts().nlargest(5).index
filtered = df[df['to'].isin(top_regions)]
sns.scatterplot(data=filtered, x='unit_cost', y='unit_price', hue='to', alpha=0.8)
plt.title(f'{plot_no}. Unit cost vs unit price for top 5 regions â€” regional pricing structure')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['qty'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of order quantities (qty) â€” volume spread and skew')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.kdeplot(df['unit_price'], fill=True)
plt.title(f'{plot_no}. Density of unit prices â€” common price points and dispersion')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_skus = df['sku'].value_counts().nlargest(6).index
sns.boxplot(data=df[df['sku'].isin(top_skus)], x='sku', y='qty')
plt.title(f'{plot_no}. Quantity distribution across top 6 SKUs â€” variability and outliers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
numeric = df.select_dtypes(include=[np.number]).copy()
corr = numeric[['qty','unit_cost','unit_price','extended_cost','extended_retail','profit','margin']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap=None)
plt.title(f'{plot_no}. Correlation matrix of numeric variables â€” relationships and multicollinearity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
monthly_price = df.groupby('month')['unit_price'].mean().reset_index()
sns.lineplot(data=monthly_price, x='month', y='unit_price', marker='o')
plt.title(f'{plot_no}. Average unit price per month â€” pricing trend over time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
pivot = df.pivot_table(index='month', columns='sku', values='extended_retail', aggfunc='sum').fillna(0)
top_skus_by_revenue = pivot.sum().nlargest(5).index
stack_data = pivot[top_skus_by_revenue]
plt.stackplot(stack_data.index, stack_data.T, labels=stack_data.columns)
plt.legend(title='SKU', bbox_to_anchor=(1.05,1), loc='upper left')
plt.title(f'{plot_no}. Revenue composition over time for top 5 SKUs â€” contribution dynamics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
df['month_of_year'] = df['month'].dt.month
monthly_season = df.groupby('month_of_year')['qty'].mean().reindex(range(1,13), fill_value=0).reset_index()
sns.barplot(data=monthly_season, x='month_of_year', y='qty')
plt.title(f'{plot_no}. Average qty by month-of-year â€” seasonal demand pattern')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='qty', y='profit', size='extended_retail', sizes=(20,200), alpha=0.6)
plt.title(f'{plot_no}. Quantity vs profit with revenue-weighted marker size â€” high-impact transactions')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
cumulative_revenue = monthly_revenue.copy()
cumulative_revenue['cumulative_revenue'] = cumulative_revenue['extended_retail'].cumsum()
sns.lineplot(data=cumulative_revenue, x='month', y='cumulative_revenue', marker='o')
plt.title(f'{plot_no}. Cumulative revenue over time â€” overall growth and inflection points')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
combo = df.groupby(['to','sku'])['extended_retail'].sum().reset_index()
top_combo = combo.sort_values('extended_retail', ascending=False).head(10)
sns.barplot(data=top_combo, x='extended_retail', y=top_combo.apply(lambda r: f"{r['to']} | {r['sku']}", axis=1))
plt.title(f'{plot_no}. Top 10 regionâ€“SKU pairs by revenue â€” key region-product drivers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
combo = df.groupby(['to','sku'])['extended_retail'].sum().reset_index()
top_combo = combo.sort_values('extended_retail', ascending=False).head(10)
sns.barplot(data=top_combo, x='extended_retail', y=top_combo.apply(lambda r: f"{r['to']} | {r['sku']}", axis=1))
plt.title(f'{plot_no}. Top 10 regionâ€“SKU pairs by revenue â€” key region-product drivers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
qty_mean = df['qty'].mean()
qty_std = df['qty'].std()
df['qty_z'] = (df['qty'] - qty_mean) / qty_std
sns.scatterplot(data=df.reset_index(), x='index', y='qty', hue=(df['qty_z'].abs() > 3), palette={False:'blue', True:'red'}, legend=False, alpha=0.7)
plt.title(f'{plot_no}. Qty outlier detection (|z|>3) â€” flagged outliers in order quantity')
show_fig()
plot_no += 1
```
