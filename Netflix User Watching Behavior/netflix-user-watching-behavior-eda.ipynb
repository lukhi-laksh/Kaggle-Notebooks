# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='gender', data=df)
plt.title(f'{plot_no}. Gender Distribution of Streaming Platform Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title(f'{plot_no}. Age Distribution of Users on the Platform')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='country', data=df, order=df['country'].value_counts().index)
plt.title(f'{plot_no}. Geographic Distribution of Users by Country')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='subscription_type', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Watch Time Variation Across Subscription Plans')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='subscription_type', data=df)
plt.title(f'{plot_no}. Popularity of Different Subscription Plans')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Average Watch Time Comparison Between Genders')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='watch_sessions_per_week', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Relationship Between Weekly Watch Sessions and Watch Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='completion_rate', y='rating_given', data=df)
plt.title(f'{plot_no}. Content Completion Rate vs User Ratings Behavior')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='primary_device', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Viewing Time Differences Across Primary Devices')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['account_age_months'], bins=20, kde=True)
plt.title(f'{plot_no}. Distribution of User Account Age in Months')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='churned', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Watch Time Comparison Between Churned and Active Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='churned', y='days_since_last_login', data=df)
plt.title(f'{plot_no}. Last Login Behavior Difference Between Churned and Active Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='favorite_genre', data=df, order=df['favorite_genre'].value_counts().index)
plt.title(f'{plot_no}. Most Preferred Content Genres Among Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='payment_method', y='monthly_fee', data=df)
plt.title(f'{plot_no}. Monthly Fee Variation Across Payment Methods')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='devices_used', y='avg_watch_time_minutes', data=df)
plt.title(f'{plot_no}. Impact of Number of Devices Used on Viewing Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='account_age_months', y='watch_sessions_per_week', data=df)
plt.title(f'{plot_no}. Relationship Between Account Age and Weekly Watch Frequency')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='subscription_type', y='completion_rate', data=df)
plt.title(f'{plot_no}. Content Completion Behavior Across Subscription Plans')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 7))
sns.heatmap(df[['age','account_age_months','monthly_fee','avg_watch_time_minutes','watch_sessions_per_week','binge_watch_sessions','completion_rate','rating_given','content_interactions','recommendation_click_rate','days_since_last_login']].corr(), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Matrix of Key User Engagement Metrics')
show_fig()
plot_no += 1
```
