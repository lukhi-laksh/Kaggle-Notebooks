# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['reels_watch_time_hours'], kde=True)
plt.title(f'{plot_no}. Distribution of Daily Reels Watch Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='daily_screen_time_hours', y='sleep_hours', data=df)
plt.title(f'{plot_no}. Screen Time vs Sleep Hours Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='reels_watch_time_hours', y='attention_span_score', data=df)
plt.title(f'{plot_no}. Reels Watching Impact on Attention Span')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='platform', y='reels_watch_time_hours', data=df)
plt.title(f'{plot_no}. Platform-wise Reels Consumption Behavior')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='stress_level', y='sleep_hours', data=df)
plt.title(f'{plot_no}. Stress Level vs Sleep Hours Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='stress_level', y='focus_level', data=df)
plt.title(f'{plot_no}. Stress Level Impact on Focus Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='attention_span_score', y='task_completion_rate', data=df)
plt.title(f'{plot_no}. Attention Span vs Task Completion Efficiency')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['sleep_hours'], kde=True)
plt.title(f'{plot_no}. Sleep Hours Distribution Among Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='daily_screen_time_hours', y='stress_level', data=df)
plt.title(f'{plot_no}. Screen Time Influence on Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='platform', y='attention_span_score', data=df)
plt.title(f'{plot_no}. Platform Impact on Attention Span Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='reels_watch_time_hours', y='stress_level', data=df)
plt.title(f'{plot_no}. Reels Watching vs Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='platform', y='task_completion_rate', data=df)
plt.title(f'{plot_no}. Platform Influence on Task Completion Rate')
show_fig()
plot_no += 1
```
