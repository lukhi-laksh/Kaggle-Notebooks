# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='position', data=df)
plt.title(f'{plot_no}. Player Distribution by Position')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='position', y='overall_rating', data=df)
plt.title(f'{plot_no}. Overall Rating Variation Across Positions')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='overall_rating', data=df)
plt.title(f'{plot_no}. Impact of Age on Overall Rating')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='matches_played', y='goals', data=df)
plt.title(f'{plot_no}. Matches Played vs Goals Scored Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='matches_played', y='assists', data=df)
plt.title(f'{plot_no}. Matches Played vs Assists Contribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['market_value_million_eur'], kde=True)
plt.title(f'{plot_no}. Market Value Distribution Among Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='injury_prone', y='minutes_played', data=df)
plt.title(f'{plot_no}. Minutes Played Comparison for Injury-Prone Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='transfer_risk_level', data=df)
plt.title(f'{plot_no}. Distribution of Transfer Risk Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='transfer_risk_level', y='market_value_million_eur', data=df)
plt.title(f'{plot_no}. Market Value Variation Across Transfer Risk Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='potential_rating', y='overall_rating', data=df)
plt.title(f'{plot_no}. Potential Rating vs Current Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='position', y='goals', data=df)
plt.title(f'{plot_no}. Average Goals Scored by Position')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='position', y='assists', data=df)
plt.title(f'{plot_no}. Average Assists by Position')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='minutes_played', y='goals', data=df)
plt.title(f'{plot_no}. Minutes Played vs Goal Output Efficiency')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='minutes_played', y='assists', data=df)
plt.title(f'{plot_no}. Minutes Played vs Assist Contribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='injury_prone', y='overall_rating', data=df)
plt.title(f'{plot_no}. Impact of Injury Proneness on Player Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['contract_years_left'], kde=True)
plt.title(f'{plot_no}. Distribution of Remaining Contract Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='contract_years_left', y='market_value_million_eur', data=df)
plt.title(f'{plot_no}. Contract Duration vs Market Value Insight')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='market_value_million_eur', data=df)
plt.title(f'{plot_no}. Age vs Market Value Trend Analysis')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Features')
show_fig()
plot_no += 1
```
