# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='fatality')
plt.title(f'{plot_no}. Fatality Outcome Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='hospitalization')
plt.title(f'{plot_no}. Hospitalization Burden Across Reported Cases')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='virus_strain', order=df['virus_strain'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Most Common Virus Strains in Reported Cases')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='transmission_type', order=df['transmission_type'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Transmission Type Pattern Across Cases')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='exposure_source', order=df['exposure_source'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Major Exposure Sources Behind Infection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='patient_age', bins=25, kde=True)
plt.title(f'{plot_no}. Patient Age Distribution and Risk Concentration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='hospitalization', y='patient_age')
plt.title(f'{plot_no}. Patient Age Difference by Hospitalization Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='fatality', y='patient_age')
plt.title(f'{plot_no}. Patient Age Difference by Fatality Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='gender', hue='hospitalization')
plt.title(f'{plot_no}. Hospitalization Comparison by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='gender', hue='fatality')
plt.title(f'{plot_no}. Fatality Comparison by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='virus_strain', hue='hospitalization', order=df['virus_strain'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Hospitalization Risk Across Virus Strains')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='virus_strain', hue='fatality', order=df['virus_strain'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Fatality Outcome Across Virus Strains')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='hospitalization', y='rodent_presence_index')
plt.title(f'{plot_no}. Rodent Presence Index by Hospitalization Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='fatality', y='rodent_presence_index')
plt.title(f'{plot_no}. Rodent Presence Index by Fatality Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='temperature_celsius', y='humidity_percent', hue='hospitalization')
plt.title(f'{plot_no}. Temperature and Humidity Relationship by Hospitalization')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='population_density', y='air_quality_index', hue='fatality')
plt.title(f'{plot_no}. Population Density and Air Quality Impact on Fatality')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='fatality', y='quarantine_days')
plt.title(f'{plot_no}. Quarantine Duration Difference by Fatality Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='country', y='rodent_presence_index', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Rodent Presence Index by Country')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Health and Environmental Risk Factors')
show_fig()
plot_no += 1
```
