# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='accident_severity', data=df)
plt.title(f'{plot_no}. Distribution of Accident Severity Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='city', data=df, order=df['city'].value_counts().index)
plt.xticks(rotation=90)
plt.title(f'{plot_no}. Accident Frequency Across Cities')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='day_of_week', data=df, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title(f'{plot_no}. Accidents by Day of Week Pattern')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='hour', data=df)
plt.title(f'{plot_no}. Hourly Distribution of Accidents')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='accident_severity', y='risk_score', data=df)
plt.title(f'{plot_no}. Risk Score Variation by Accident Severity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['risk_score'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Risk Scores')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='traffic_density', hue='accident_severity', data=df)
plt.title(f'{plot_no}. Impact of Traffic Density on Accident Severity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='visibility', hue='accident_severity', data=df)
plt.title(f'{plot_no}. Visibility Conditions vs Accident Severity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='temperature', y='risk_score', hue='accident_severity', data=df)
plt.title(f'{plot_no}. Temperature Influence on Risk Score and Severity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='is_weekend', y='risk_score', data=df)
plt.title(f'{plot_no}. Weekend vs Weekday Risk Score Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='cause', data=df, order=df['cause'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Major Causes of Accidents')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='cause', hue='accident_severity', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Accident Causes by Severity Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='accident_severity', y='vehicles_involved', data=df)
plt.title(f'{plot_no}. Vehicles Involved Across Severity Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='accident_severity', y='casualties', data=df)
plt.title(f'{plot_no}. Casualties Distribution by Severity Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='is_peak_hour', hue='accident_severity', data=df)
plt.title(f'{plot_no}. Peak Hour Influence on Accident Severity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Features')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='traffic_density', y='risk_score', data=df)
plt.title(f'{plot_no}. Risk Score Distribution Across Traffic Density Levels')
show_fig()
plot_no += 1
```
