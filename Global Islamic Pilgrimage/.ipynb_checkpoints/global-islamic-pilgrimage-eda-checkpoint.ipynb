# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Total_Pilgrims', estimator='sum', errorbar=None, marker='o')
plt.title(f'{plot_no}. Yearly Growth Trend of Total Pilgrims')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Country', y='Total_Pilgrims', estimator='sum', errorbar=None)
plt.xticks(rotation=908)
plt.title(f'{plot_no}. Country-Wise Total Pilgrim Contribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Pilgrimage_Type', y='Total_Pilgrims', estimator='sum', errorbar=None)
plt.title(f'{plot_no}. Total Pilgrims by Pilgrimage Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Total_Pilgrims', hue='Pilgrimage_Type', estimator='sum', errorbar=None, marker='o')
plt.title(f'{plot_no}. Pilgrimage Type Demand Trend Over Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Economic_Impact_USD', estimator='sum', errorbar=None, marker='o')
plt.title(f'{plot_no}. Yearly Economic Impact Generated by Pilgrimage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Country', y='Economic_Impact_USD', estimator='sum', errorbar=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Country-Wise Total Economic Impact')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Average_Travel_Cost_USD', y='Economic_Impact_USD', hue='Pilgrimage_Type', size='Total_Pilgrims', sizes=(30, 250))
plt.title(f'{plot_no}. Travel Cost Impact on Economic Contribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Pilgrimage_Type', y='Average_Travel_Cost_USD')
plt.title(f'{plot_no}. Travel Cost Distribution Across Pilgrimage Types')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Country', y='Satisfaction_Rating')
plt.xticks(rotation=90)
plt.title(f'{plot_no}. Country-Wise Satisfaction Rating Variation')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Pilgrimage_Type', y='Satisfaction_Rating')
plt.title(f'{plot_no}. Satisfaction Rating Distribution by Pilgrimage Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Flights_Arranged', y='Hotels_Booked', hue='Pilgrimage_Type', size='Total_Pilgrims', sizes=(30, 250))
plt.title(f'{plot_no}. Flight and Hotel Capacity Planning Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Average_Age', hue='Pilgrimage_Type', kde=True, bins=20)
plt.title(f'{plot_no}. Age Distribution of Pilgrims by Pilgrimage Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Male_Percentage', y='Female_Percentage', hue='Pilgrimage_Type', size='Total_Pilgrims', sizes=(30, 250))
plt.title(f'{plot_no}. Gender Percentage Balance Across Pilgrimage Types')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Satisfaction_Rating', hue='Pilgrimage_Type', estimator='mean', errorbar=None, marker='o')
plt.title(f'{plot_no}. Average Satisfaction Rating Trend by Pilgrimage Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title(f'{plot_no}. Correlation Heatmap of Pilgrimage Dataset Metrics')
show_fig()
plot_no += 1
```
