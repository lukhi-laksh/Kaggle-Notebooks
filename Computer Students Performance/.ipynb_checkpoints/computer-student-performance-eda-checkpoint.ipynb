# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Gender', data=df)
plt.title(f'{plot_no}. Gender Distribution of Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['GPA'], kde=True, bins=8)
plt.title(f'{plot_no}. GPA Distribution and Academic Performance Spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Gender', y='GPA', data=df)
plt.title(f'{plot_no}. GPA Comparison Across Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='GPA', hue='Gender', data=df)
plt.title(f'{plot_no}. Relationship Between Age and GPA by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='Future Career', data=df, order=df['Future Career'].value_counts().index)
plt.title(f'{plot_no}. Future Career Aspirations Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Python', data=df, order=df['Python'].value_counts().index)
plt.title(f'{plot_no}. Python Skill Level Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='SQL', data=df, order=df['SQL'].value_counts().index)
plt.title(f'{plot_no}. SQL Skill Level Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Java', data=df, order=df['Java'].value_counts().index)
plt.title(f'{plot_no}. Java Skill Level Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Future Career', hue='Gender', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Career Aspirations by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Interested Domain', hue='Gender', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Interested Domain Preference by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Python', y='GPA', data=df, order=['Weak','Average','Strong'])
plt.title(f'{plot_no}. Impact of Python Skill Level on GPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='SQL', y='GPA', data=df, order=['Weak','Average','Strong'])
plt.title(f'{plot_no}. Impact of SQL Skill Level on GPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Java', y='GPA', data=df, order=['Weak','Average','Strong'])
plt.title(f'{plot_no}. Impact of Java Skill Level on GPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Projects', data=df, order=df['Projects'].value_counts().index)
plt.xticks(rotation=90)
plt.title(f'{plot_no}. Project Types Distribution Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Gender', y='Age', data=df)
plt.title(f'{plot_no}. Age Distribution Spread by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='GPA', style='Python', data=df)
plt.title(f'{plot_no}. Age vs GPA Pattern by Python Skill Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
skill_map = df[['Python','SQL','Java']].replace({'Weak':1,'Average':2,'Strong':3})
sns.heatmap(skill_map.corr(), annot=True)
plt.title(f'{plot_no}. Correlation Between Programming Skill Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Interested Domain', y='GPA', data=df, estimator=np.mean)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average GPA Across Different Interested Domains')
show_fig()
plot_no += 1
```
