# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Purchase Amount (USD)'].dropna(), kde=True)
plt.title(f'{plot_no}. Distribution of Purchase Amounts (USD) â€” central tendency and spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Category', y='Purchase Amount (USD)', data=df)
plt.title(f'{plot_no}. Category-wise spread of Purchase Amounts to identify high-spend categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='Category', data=df, order=df['Category'].value_counts().index)
plt.title(f'{plot_no}. Frequency of Purchases by Category â€” where most transactions occur')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='Purchase Amount (USD)', data=df, estimator=np.mean)
plt.title(f'{plot_no}. Average Purchase Amount by Gender â€” comparison of mean spend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x='Age', y='Purchase Amount (USD)', data=df, scatter_kws={'s':20}, lowess=True)
plt.title(f'{plot_no}. Relationship between Age and Purchase Amount â€” spending trend with age')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
num_cols = ['Age','Purchase Amount (USD)','Review Rating','Previous Purchases']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', vmin=-1, vmax=1)
plt.title(f'{plot_no}. Correlation heatmap of numeric customer and purchase metrics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Shipping Type', y='Purchase Amount (USD)', data=df)
plt.title(f'{plot_no}. How Purchase Amounts vary by Shipping Type â€” cost vs delivery option')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(y='Frequency of Purchases', data=df, order=df['Frequency of Purchases'].value_counts().index)
plt.title(f'{plot_no}. Distribution of Purchase Frequencies â€” customer buying cadence')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Subscription Status', y='Purchase Amount (USD)', data=df)
plt.title(f'{plot_no}. Purchase Amount distributions by Subscription Status â€” subscribers vs non-subscribers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Subscription Status', y='Purchase Amount (USD)', data=df)
plt.title(f'{plot_no}. Purchase Amount distributions by Subscription Status â€” subscribers vs non-subscribers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
avg_rating = df.groupby('Category')['Review Rating'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='Review Rating', y='Category', data=avg_rating)
plt.title(f'{plot_no}. Average Review Rating by Category â€” categories with higher satisfaction')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
ct = pd.crosstab(df['Category'], df['Gender'])
ct.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title(f'{plot_no}. Category composition by Gender (stacked) â€” gender mix per category')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Previous Purchases', hue='Subscription Status', data=df)
plt.title(f'{plot_no}. Age vs Previous Purchases colored by Subscription Status â€” loyalty patterns by age')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Season', y='Purchase Amount (USD)', data=df, order=['Spring','Summer','Fall','Winter'] if set(['Spring','Summer','Fall','Winter']).issubset(set(df['Season'].unique())) else None)
plt.title(f'{plot_no}. Seasonal variation in Purchase Amounts â€” which seasons drive higher spend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_locations = df.groupby('Location')['Purchase Amount (USD)'].sum().nlargest(10).reset_index()
sns.barplot(x='Purchase Amount (USD)', y='Location', data=top_locations)
plt.title(f'{plot_no}. Top 10 Locations by Total Purchase Amount â€” geographic revenue hotspots')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Payment Method', y='Purchase Amount (USD)', data=df, estimator=np.mean, order=df.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False).index)
plt.title(f'{plot_no}. Average Purchase Amount by Payment Method â€” payment channels with higher average spend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Promo Code Used', y='Purchase Amount (USD)', data=df)
plt.title(f'{plot_no}. Effect of Promo Code Usage on Purchase Amounts â€” promotional lift or discount effect')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
plt.hexbin(df['Purchase Amount (USD)'].dropna(), df['Review Rating'].dropna(), gridsize=30, cmap='Blues')
plt.xlabel('Purchase Amount (USD)')
plt.ylabel('Review Rating')
plt.title(f'{plot_no}. Density of Purchase Amount vs Review Rating â€” high-density spending & satisfaction areas')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
ct_color = df.groupby(['Category','Color']).size().reset_index(name='count')
top_colors = ct_color.sort_values('count', ascending=False).groupby('Category').head(3)
sns.catplot(x='Category', y='count', hue='Color', data=top_colors, kind='bar', height=6, aspect=1.5)
plt.title(f'{plot_no}. Top 3 Colors per Category by count â€” color preferences by category')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
df['AgeBin'] = pd.cut(df['Age'], bins=[0,18,30,45,60,120], labels=['0-18','19-30','31-45','46-60','60+'])
sns.boxplot(x='AgeBin', y='Previous Purchases', data=df)
plt.title(f'{plot_no}. Previous Purchases distribution across Age Groups â€” purchase history by age segment')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.pairplot(df[['Age','Purchase Amount (USD)','Review Rating','Previous Purchases']].dropna())
plt.suptitle(f'{plot_no}. Pairwise relationships between Age, Purchase Amount, Review Rating, and Previous Purchases', y=1.02)
show_fig()
plot_no += 1
```
