# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Attack Type', order=df['Attack Type'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Most Frequent Cyber Attack Types Across All Countries')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Target Industry', order=df['Target Industry'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Industries Most Targeted by Cyber Attacks')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Country', y='Financial Loss (in Million $)', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Total Financial Loss by Country Due to Cyber Attacks')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.groupby('Year')['Financial Loss (in Million $)'].sum().reset_index(), x='Year', y='Financial Loss (in Million $)')
plt.title(f'{plot_no}. Yearly Trend of Total Financial Loss from Cyber Attacks')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Attack Type', y='Financial Loss (in Million $)')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Financial Loss Distribution by Attack Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Financial Loss (in Million $)'], kde=True, bins=30)
plt.title(f'{plot_no}. Distribution of Financial Loss Across Incidents')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Security Vulnerability Type', y='Financial Loss (in Million $)', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Total Financial Impact by Vulnerability Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Defense Mechanism Used', y='Incident Resolution Time (in Hours)')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Incident Resolution Time by Defense Mechanism')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Attack Source', y='Financial Loss (in Million $)', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Financial Loss Contribution by Attack Source')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Year')
plt.title(f'{plot_no}. Annual Frequency of Cyber Security Incidents')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Attack Type', y='Incident Resolution Time (in Hours)')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Resolution Time Variability by Attack Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Target Industry', y='Number of Affected Users', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Total Users Impacted per Industry')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
pivot_df = df.pivot_table(values='Financial Loss (in Million $)', index='Country', columns='Attack Type', aggfunc='sum')
sns.heatmap(pivot_df, annot=True, cmap='viridis')
plt.title(f'{plot_no}. Heatmap of Financial Loss by Country and Attack Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Incident Resolution Time (in Hours)', y='Financial Loss (in Million $)', hue='Attack Source')
plt.title(f'{plot_no}. Impact of Resolution Time on Financial Loss by Attack Source')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Security Vulnerability Type', y='Number of Affected Users')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Users Affected Distribution by Vulnerability Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.groupby('Year')['Number of Affected Users'].sum().reset_index(), x='Year', y='Number of Affected Users')
plt.title(f'{plot_no}. Yearly Trend of Total Affected Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Defense Mechanism Used', y='Financial Loss (in Million $)', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Financial Loss by Defense Mechanism Effectiveness')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr_df = df[['Financial Loss (in Million $)', 'Number of Affected Users', 'Incident Resolution Time (in Hours)']].corr()
sns.heatmap(corr_df, annot=True, cmap='coolwarm', fmt='.2f')
plt.title(f'{plot_no}. Correlation Matrix of Key Impact Metrics')
show_fig()
plot_no += 1
```
