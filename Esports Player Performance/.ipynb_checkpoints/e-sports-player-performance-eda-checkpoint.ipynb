# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='match_outcome', hue='match_outcome')
plt.title(f'{plot_no}. Overall Match Outcome Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='match_outcome', y='performance_score')
plt.title(f'{plot_no}. Performance Score Distribution by Match Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='kills', y='performance_score', hue='match_outcome')
plt.title(f'{plot_no}. Impact of Kills on Performance Score by Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='player_role', y='kills')
plt.title(f'{plot_no}. Kill Distribution Across Player Roles')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='match_type', y='performance_score')
plt.title(f'{plot_no}. Performance Score Variation by Match Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='reaction_time_ms', y='accuracy_percent', hue='match_outcome')
plt.title(f'{plot_no}. Reaction Time vs Accuracy by Match Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='mvp_award', y='performance_score')
plt.title(f'{plot_no}. Performance Score Comparison for MVP vs Non-MVP Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='team_name', y='performance_score', estimator=np.mean)
plt.title(f'{plot_no}. Average Team Performance Score Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='fatigue_index', y='performance_score', hue='match_outcome')
plt.title(f'{plot_no}. Fatigue Index Impact on Performance Score')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['accuracy_percent'], kde=True)
plt.title(f'{plot_no}. Distribution of Player Accuracy Percentage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['reaction_time_ms'], kde=True)
plt.title(f'{plot_no}. Distribution of Player Reaction Time (ms)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='map_played', y='kills')
plt.title(f'{plot_no}. Kill Performance Across Different Maps')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='match_outcome', y='fatigue_index')
plt.title(f'{plot_no}. Fatigue Index Comparison Between Wins and Losses')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df[['kills','assists','deaths','accuracy_percent','reaction_time_ms','fatigue_index','performance_score','win_probability']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Key Performance Metrics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='player_role', y='accuracy_percent')
plt.title(f'{plot_no}. Accuracy Percentage Variation by Player Role')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='match_type', y='win_probability', estimator=np.mean)
plt.title(f'{plot_no}. Average Win Probability by Match Type')
show_fig()
plot_no += 1
```
