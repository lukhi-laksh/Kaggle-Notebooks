# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['overall_rating'], kde=True, bins=20)
plt.title(f'{plot_no}. Overall Rating Distribution of Football Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['potential_rating'], kde=True, bins=20)
plt.title(f'{plot_no}. Potential Rating Distribution Across Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='overall_rating', y='market_value_million_eur', hue='transfer_risk_level')
plt.title(f'{plot_no}. Overall Rating Impact on Market Value by Transfer Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='potential_rating', y='market_value_million_eur', hue='injury_prone')
plt.title(f'{plot_no}. Potential Rating vs Market Value by Injury Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='transfer_risk_level', y='market_value_million_eur')
plt.title(f'{plot_no}. Market Value Comparison Across Transfer Risk Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='injury_prone', y='overall_rating')
plt.title(f'{plot_no}. Overall Rating Difference Between Injury-Prone and Fit Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='position', order=df['position'].value_counts().index)
plt.title(f'{plot_no}. Player Count Distribution by Position')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='position', y='goals', estimator='mean')
plt.title(f'{plot_no}. Average Goals Scored by Player Position')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='position', y='assists', estimator='mean')
plt.title(f'{plot_no}. Average Assists Created by Player Position')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='club', y='overall_rating')
plt.title(f'{plot_no}. Overall Rating Spread Across Clubs')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='club', y='market_value_million_eur', estimator='mean')
plt.title(f'{plot_no}. Average Market Value by Club')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='nationality', order=df['nationality'].value_counts().index)
plt.title(f'{plot_no}. Player Distribution by Nationality')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='nationality', y='overall_rating', estimator='mean')
plt.title(f'{plot_no}. Average Overall Rating by Nationality')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='transfer_risk_level', y='contract_years_left')
plt.title(f'{plot_no}. Contract Years Left Across Transfer Risk Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='contract_years_left', y='market_value_million_eur', hue='transfer_risk_level')
plt.title(f'{plot_no}. Contract Length vs Market Value by Transfer Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Football Player Features')
show_fig()
plot_no += 1
```
