# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['daily_screen_time_hours'], kde=True)
plt.title(f'{plot_no}. Distribution of Daily Screen Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='daily_screen_time_hours', y='productivity_score', data=df)
plt.title(f'{plot_no}. Screen Time vs Productivity Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='social_media_hours', y='stress_level', data=df)
plt.title(f'{plot_no}. Social Media Usage Impact on Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='platform', y='daily_screen_time_hours', data=df)
plt.title(f'{plot_no}. Platform-wise Screen Time Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='platform', y='productivity_score', data=df)
plt.title(f'{plot_no}. Productivity Distribution Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sleep_hours', y='stress_level', data=df)
plt.title(f'{plot_no}. Sleep Duration vs Stress Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='exercise_minutes', y='stress_level', data=df)
plt.title(f'{plot_no}. Exercise Impact on Stress Reduction')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='study_work_hours', y='productivity_score', data=df)
plt.title(f'{plot_no}. Work/Study Hours vs Productivity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['stress_level'], kde=True)
plt.title(f'{plot_no}. Stress Level Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['productivity_score'], kde=True)
plt.title(f'{plot_no}. Productivity Score Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='platform', y='stress_level', data=df)
plt.title(f'{plot_no}. Average Stress Levels by Platform')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='platform', y='productivity_score', data=df)
plt.title(f'{plot_no}. Average Productivity by Platform')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='daily_screen_time_hours', y='sleep_hours', data=df)
plt.title(f'{plot_no}. Screen Time vs Sleep Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='social_media_hours', y='productivity_score', data=df)
plt.title(f'{plot_no}. Social Media Usage vs Productivity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='platform', y='sleep_hours', data=df)
plt.title(f'{plot_no}. Sleep Patterns Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='exercise_minutes', y='productivity_score', data=df)
plt.title(f'{plot_no}. Exercise vs Productivity Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='stress_level', data=df)
plt.title(f'{plot_no}. Age vs Stress Level Trend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='daily_screen_time_hours', data=df)
plt.title(f'{plot_no}. Age vs Screen Time Behavior')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 9))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of All Numerical Features')
show_fig()
plot_no += 1
```
