# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['weekday_screen_hours'], bins=30, kde=True)
plt.title(f'{plot_no}. Weekday Screen Hours Distribution Shows Daily Usage Pattern')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['weekend_screen_hours'], bins=30, kde=True)
plt.title(f'{plot_no}. Weekend Screen Hours Distribution Shows Higher Free-Time Usage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df[['weekday_screen_hours', 'weekend_screen_hours']])
plt.title(f'{plot_no}. Weekday vs Weekend Screen Time Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='age_group', y='weekday_screen_hours')
plt.title(f'{plot_no}. Age Group Wise Weekday Screen Time Difference')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='platform', y='weekday_screen_hours', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Weekday Screen Time by Platform')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='platform', y='focus_span_minutes')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Platform Wise Focus Span Variation')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='weekday_screen_hours', y='focus_span_minutes', hue='gender')
plt.title(f'{plot_no}. Weekday Screen Time Impact on Focus Span by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='gender', y='physical_activity_hours_weekly')
plt.title(f'{plot_no}. Gender Wise Weekly Physical Activity Difference')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='age_group', y='physical_activity_hours_weekly', estimator='mean')
plt.title(f'{plot_no}. Average Weekly Physical Activity by Age Group')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.groupby('year')[['weekday_screen_hours', 'weekend_screen_hours']].mean().reset_index(), x='year', y='weekday_screen_hours', label='Weekday')
sns.lineplot(data=df.groupby('year')[['weekday_screen_hours', 'weekend_screen_hours']].mean().reset_index(), x='year', y='weekend_screen_hours', label='Weekend')
plt.title(f'{plot_no}. Year Wise Screen Time Trend for Weekday and Weekend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='country', y='focus_span_minutes', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Country Wise Average Focus Span Comparison')
show_fig()
plot_no += 1
```

```python
corr = df.select_dtypes(include='number').corr()
fig = plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap Shows Strongest Numerical Feature Relationships')
show_fig()
plot_no += 1
```
