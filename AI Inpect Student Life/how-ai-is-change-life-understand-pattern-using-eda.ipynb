# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Students Using AI Tools')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Major', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Distribution of Students Across Different Majors')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Primary_AI_Tool', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Popularity of AI Tools Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Primary_AI_Tool', y='GPA_Post_AI', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. GPA After AI Usage by Tool Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='GPA_Baseline', y='GPA_Post_AI', data=df)
plt.title(f'{plot_no}. Baseline GPA vs Post-AI GPA Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Time_Saved_Hours_Weekly'], kde=True)
plt.title(f'{plot_no}. Weekly Time Saved Using AI Tools Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Time_Saved_Hours_Weekly', y='GPA_Post_AI', data=df)
plt.title(f'{plot_no}. Impact of Time Saved on Post-AI GPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Task_Frequency_Daily', data=df)
plt.title(f'{plot_no}. Daily Frequency of AI Tool Usage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Task_Frequency_Daily', y='GPA_Post_AI', data=df)
plt.title(f'{plot_no}. Effect of AI Usage Frequency on Academic Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Main_Usage_Case', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Primary Use Cases of AI Tools by Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Main_Usage_Case', y='Time_Saved_Hours_Weekly', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Time Saved Across Different AI Usage Cases')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='AI_Ethics_Concern', data=df)
plt.title(f'{plot_no}. Distribution of AI Ethics Concern Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='AI_Ethics_Concern', y='Career_Confidence_Score', data=df)
plt.title(f'{plot_no}. Ethics Concerns vs Career Confidence')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Career_Confidence_Score', data=df)
plt.title(f'{plot_no}. Age vs Career Confidence Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Task_Frequency_Daily', y='Time_Saved_Hours_Weekly', data=df)
plt.title(f'{plot_no}. AI Usage Frequency vs Time Saved')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Major', y='GPA_Post_AI', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Post-AI GPA Variation Across Majors')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Primary_AI_Tool', y='Time_Saved_Hours_Weekly', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Distribution of Time Saved by AI Tool')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='GPA_Baseline', y='Time_Saved_Hours_Weekly', data=df)
plt.title(f'{plot_no}. Baseline GPA vs Time Saved Using AI')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='AI_Ethics_Concern', y='GPA_Post_AI', data=df)
plt.title(f'{plot_no}. Ethics Concern Impact on Academic Outcomes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Features')
show_fig()
plot_no += 1
```
