# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Post_Semester_GPA'], kde=True)
plt.title(f'{plot_no}. Post Semester GPA Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Pre_Semester_GPA', y='Post_Semester_GPA', hue='Burnout_Risk_Level')
plt.title(f'{plot_no}. Pre GPA vs Post GPA by Burnout Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Burnout_Risk_Level', y='Post_Semester_GPA')
plt.title(f'{plot_no}. Post GPA Comparison Across Burnout Risk Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Major_Category', y='Post_Semester_GPA')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Post GPA Variation by Major Category')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Year_of_Study', y='Post_Semester_GPA', estimator='mean')
plt.title(f'{plot_no}. Average Post GPA by Year of Study')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Prompt_Engineering_Skill', y='Post_Semester_GPA', estimator='mean')
plt.title(f'{plot_no}. Average Post GPA by Prompt Engineering Skill')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Weekly_GenAI_Hours', y='Traditional_Study_Hours', hue='Burnout_Risk_Level')
plt.title(f'{plot_no}. GenAI Hours vs Traditional Study Hours by Burnout Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Primary_Use_Case', y='Weekly_GenAI_Hours')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. GenAI Usage Intensity by Primary Use Case')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Burnout_Risk_Level', hue='Paid_Subscription')
plt.title(f'{plot_no}. Burnout Risk Count by Paid Subscription Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Institutional_Policy', y='Post_Semester_GPA')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Post GPA Across Institutional AI Policy Types')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Perceived_AI_Dependency', y='Skill_Retention_Score', hue='Prompt_Engineering_Skill')
plt.title(f'{plot_no}. AI Dependency vs Skill Retention by Prompt Skill')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Weekly_GenAI_Hours', y='Skill_Retention_Score', hue='Burnout_Risk_Level')
plt.title(f'{plot_no}. GenAI Hours vs Skill Retention by Burnout Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Primary_Use_Case', y='Skill_Retention_Score', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Skill Retention by Primary AI Use Case')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Tool_Diversity', y='Post_Semester_GPA')
plt.title(f'{plot_no}. Post GPA Distribution by AI Tool Diversity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Major_Category', hue='Burnout_Risk_Level')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Burnout Risk Pattern Across Major Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Numeric Feature Correlation Heatmap')
show_fig()
plot_no += 1
```
