# EDA

```python
def show_fig():
        plt.tight_layout()
        plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title(f'{plot_no}. Distribution of Restaurant Ratings ‚Äî shape, central tendency, and spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Michelin_Stars', y='Average_Price_USD', data=df)
plt.title(f'{plot_no}. Average Price by Michelin Stars ‚Äî price dispersion across star levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Continent', y='Average_Price_USD', data=df, inner='quartile')
plt.title(f'{plot_no}. Price distribution by Continent ‚Äî regional pricing structure and spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_cuisines = df['Cuisine_Type'].value_counts().nlargest(10).index
sns.countplot(y='Cuisine_Type', data=df[df['Cuisine_Type'].isin(top_cuisines)], order=top_cuisines)
plt.title(f'{plot_no}. Top 10 Cuisine Types by Count ‚Äî popularity of cuisine styles')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
num_cols = ['Michelin_Stars','Average_Price_USD','Seating_Capacity','Rating','Years_Operating','Established_Year']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', vmin=-1, vmax=1, square=True)
plt.title(f'{plot_no}. Correlation Matrix ‚Äî numeric feature relationships')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Established_Year', y='Average_Price_USD', hue='Star_Category', alpha=0.9)
plt.title(f'{plot_no}. Established Year vs Average Price ‚Äî pricing trends by star category over time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Star_Category', y='Seating_Capacity', data=df)
plt.title(f'{plot_no}. Seating Capacity Distribution by Star Category ‚Äî size profile of restaurants by prestige')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
for cat in df['Star_Category'].unique():
    subset = df[df['Star_Category']==cat]
    sns.kdeplot(subset['Average_Price_USD'], label=str(cat), common_norm=False)
plt.legend(title='Star Category')
plt.title(f'{plot_no}. KDE of Average Price by Star Category ‚Äî comparative price density')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Average_Price_USD', y='Rating', hue='Michelin_Stars', alpha=0.8)
sns.regplot(data=df, x='Average_Price_USD', y='Rating', scatter=False, lowess=True)
plt.title(f'{plot_no}. Price vs Rating with Trend ‚Äî does higher price predict higher rating?')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_cities = df.groupby('City')['Average_Price_USD'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title(f'{plot_no}. Top 10 Cities by Mean Average Price ‚Äî most expensive dining cities in dataset')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Price_Category', y='Rating', data=df, order=df['Price_Category'].value_counts().index)
plt.title(f'{plot_no}. Rating Distribution across Price Categories ‚Äî value vs perceived quality')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Michelin_Stars', hue='Continent', data=df)
plt.title(f'{plot_no}. Michelin Stars Count by Continent ‚Äî geographic distribution of starred restaurants')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Years_Operating', y='Average_Price_USD', size='Rating', hue='Continent', sizes=(50,300), alpha=0.8)
plt.xlim(left=0)
plt.title(f'{plot_no}. Years Operating vs Price colored by Continent (bubble=Rating) ‚Äî longevity, price and regional effects')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_cuisines_capacity = df.groupby('Cuisine_Type')['Seating_Capacity'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=top_cuisines_capacity.values, y=top_cuisines_capacity.index)
plt.title(f'{plot_no}. Mean Seating Capacity for Top 10 Cuisine Types ‚Äî which cuisines run larger venues')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Established_Year'].dropna(), bins=20, kde=False)
plt.axvline(df['Established_Year'].median(), linestyle='--')
plt.title(f'{plot_no}. Distribution of Established Year with Median ‚Äî eras when these restaurants were founded')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
pivot = df.pivot_table(index='Continent', columns='Star_Category', values='Rating', aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt='.2f', cbar_kws={'label':'Mean Rating'})
plt.title(f'{plot_no}. Mean Rating by Continent and Star Category ‚Äî regional quality differences by prestige')
show_fig()
plot_no += 1
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, balanced_accuracy_score, f1_score
import numpy as np
import matplotlib.pyplot as plt

target = 'Michelin_Stars'
drop_cols = ['Restaurant_Name', 'Chef', 'Signature_Dish', 'City']  # drop identifiers and textual noise
df_model = df.copy()
df_model = df_model.drop(columns=[c for c in drop_cols if c in df_model.columns])

y = df_model[target].astype(int)
X = df_model.drop(columns=[target])

numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object', 'category']).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='__missing__')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
], remainder='drop', verbose_feature_names_out=False)

clf = RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced')

pipe = Pipeline(steps=[('preproc', preprocessor), ('clf', clf)])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)
y_proba = pipe.predict_proba(X_test) if hasattr(pipe.named_steps['clf'], "predict_proba") else None

print("Cross-validated balanced accuracy (5-fold stratified):")
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(pipe, X, y, cv=cv, scoring='balanced_accuracy')
print(f"Balanced accuracy per fold: {np.round(cv_scores,3)}")
print(f"Mean balanced accuracy: {cv_scores.mean():.3f}\n")

print("Test set evaluation:")
print(f"Balanced accuracy: {balanced_accuracy_score(y_test, y_pred):.3f}")
print(f"Macro F1 score: {f1_score(y_test, y_pred, average='macro'):.3f}\n")
print("Classification report (test set):")
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred, labels=pipe.named_steps['clf'].classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=pipe.named_steps['clf'].classes_)
fig = plt.figure(figsize=(8,6))
disp.plot(ax=plt.gca())
plt.title('Confusion Matrix: Michelin Stars (test set)')
plt.tight_layout()
plt.show()

preproc = pipe.named_steps['preproc']
onehot_columns = []
if hasattr(preproc, 'transformers_'):
    # Build feature names from ColumnTransformer (sklearn >=1.0 style)
    feature_names = []
    for name, trans, cols in preproc.transformers_:
        if name == 'remainder':
            continue
        if name == 'num':
            feature_names.extend(cols)
        elif name == 'cat':
            ohe = trans.named_steps['onehot']
            cats = ohe.get_feature_names_out(cols)
            feature_names.extend(cats)
else:
    feature_names = numeric_features + categorical_features

importances = pipe.named_steps['clf'].feature_importances_
idx = np.argsort(importances)[::-1][:20]
top_feats = [(feature_names[i], importances[i]) for i in idx]

fig = plt.figure(figsize=(10,6))
feat_names_plot = [f'{n} ({imp:.3f})' for n, imp in top_feats[::-1]]
vals = [imp for n, imp in top_feats[::-1]]
plt.barh(feat_names_plot, vals)
plt.title('Top 20 Feature Importances for predicting Michelin_Stars')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```
