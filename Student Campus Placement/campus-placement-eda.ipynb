# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Salary_LPA', kde=True)
plt.title(f'{plot_no}. Salary LPA Distribution Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='College_Tier', y='Salary_LPA')
plt.title(f'{plot_no}. College Tier Impact on Salary LPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Specialization', y='Salary_LPA')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Salary Comparison Across Specializations')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='CGPA', y='Salary_LPA')
plt.title(f'{plot_no}. CGPA Relationship With Salary and Placement')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Internships', y='Salary_LPA')
plt.title(f'{plot_no}. Internship Count Impact on Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Projects_Count', y='Salary_LPA')
plt.title(f'{plot_no}. Project Count Relationship With Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='GitHub_Contributions', y='Salary_LPA')
plt.title(f'{plot_no}. GitHub Contributions Impact on Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Communication_Skills', y='Mock_Interview_Score')
plt.title(f'{plot_no}. Communication and Mock Interview Score Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Resume_Score', y='Salary_LPA')
plt.title(f'{plot_no}. Resume Score Impact on Salary LPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Aptitude_Test_Score', y='Salary_LPA')
plt.title(f'{plot_no}. Aptitude Score Relationship With Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='AI_ML_Skill_Level', y='Salary_LPA')
plt.title(f'{plot_no}. AI ML Skill Level Impact on Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 9))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Student Placement Features')
show_fig()
plot_no += 1
```
