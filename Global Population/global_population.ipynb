# EDA

## Univariate

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Loop through each numeric column and plot histogram
for col in numeric_cols:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Univariate Analysis - {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

## Bivariate

```python
pairs = [
    ('Population(in millions)', 'Population density'),
    ('Male Population(in millions)', 'Female Population(in millions)'),
    ('Population Aged 0 to 14 (%)', 'Population Aged 60 and Over (%)'),
    ('Population(in millions)', 'Sex ratio (males per 100 females)'),
    ('Population density', 'Sex ratio (males per 100 females)')
]

for col_x, col_y in pairs:
    plt.figure(figsize=(7,4))
    sns.scatterplot(x=df[col_x], y=df[col_y], hue=df['Country'], palette='Set2', s=100)
    plt.title(f"{col_x} vs {col_y}")
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.legend([],[], frameon=False)
    plt.show()
```
