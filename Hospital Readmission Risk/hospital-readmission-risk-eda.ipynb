# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Readmitted_Within_30_Days')
plt.title(f'{plot_no}. 30-Day Readmission Class Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', hue='Readmitted_Within_30_Days', kde=True, bins=30)
plt.title(f'{plot_no}. Age Distribution by 30-Day Readmission Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Length_of_Stay')
plt.title(f'{plot_no}. Length of Stay Difference Between Readmitted and Non-Readmitted Patients')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Socioeconomic_Risk_Score')
plt.title(f'{plot_no}. Socioeconomic Risk Score Impact on Readmission')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Previous_Admissions_6M')
plt.title(f'{plot_no}. Previous 6-Month Admissions vs Readmission Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Previous_Readmissions_1Y')
plt.title(f'{plot_no}. Past 1-Year Readmissions as a Strong Readmission Indicator')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Medication_Adherence_Score', hue='Readmitted_Within_30_Days', kde=True, bins=30)
plt.title(f'{plot_no}. Medication Adherence Score Distribution by Readmission')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Number_of_Medications')
plt.title(f'{plot_no}. Medication Burden Comparison by Readmission Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Readmitted_Within_30_Days', y='Medication_Change_Count')
plt.title(f'{plot_no}. Medication Change Count and Readmission Risk')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='High_Risk_Medication_Flag', hue='Readmitted_Within_30_Days')
plt.title(f'{plot_no}. High-Risk Medication Flag vs Readmission Frequency')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Followup_Appointment_Scheduled', hue='Readmitted_Within_30_Days')
plt.title(f'{plot_no}. Follow-Up Appointment Scheduling and Readmission Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Admission_Type', hue='Readmitted_Within_30_Days')
plt.title(f'{plot_no}. Admission Type Patterns Across Readmission Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Primary_Diagnosis_Group', hue='Readmitted_Within_30_Days')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Primary Diagnosis Group and Readmission Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Discharge_Disposition', hue='Readmitted_Within_30_Days')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Discharge Disposition Influence on Readmission')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Insurance_Type', hue='Readmitted_Within_30_Days')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Insurance Type Comparison by Readmission Status')
show_fig()
plot_no += 1
```

```python
corr = df.select_dtypes(include='number').corr()
fig = plt.figure(figsize=(14,10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title(f'{plot_no}. Correlation Heatmap of Numeric Health and Readmission Features')
show_fig()
plot_no += 1
```
