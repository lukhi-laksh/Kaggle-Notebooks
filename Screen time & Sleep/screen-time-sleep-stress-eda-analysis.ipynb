# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Daily_Phone_Hours'], kde=True)
plt.title(f'{plot_no}. Distribution of Daily Phone Usage Hours Across Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Sleep_Hours'], kde=True)
plt.title(f'{plot_no}. Distribution of Sleep Hours: Identifying Sleep Patterns')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Stress_Level'], kde=True)
plt.title(f'{plot_no}. Stress Level Distribution Among Smartphone Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Daily_Phone_Hours', y='Sleep_Hours', data=df, hue='Stress_Level')
plt.title(f'{plot_no}. Impact of Phone Usage on Sleep Duration and Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Social_Media_Hours', y='Work_Productivity_Score', data=df, hue='Gender')
plt.title(f'{plot_no}. Social Media Usage vs Work Productivity by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Gender', y='Daily_Phone_Hours', data=df)
plt.title(f'{plot_no}. Comparison of Daily Phone Usage Between Genders')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Device_Type', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Stress Level Comparison Across Device Types')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Occupation', y='Work_Productivity_Score', data=df)
plt.title(f'{plot_no}. Productivity Score Distribution Across Occupations')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Caffeine_Intake_Cups', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Relationship Between Caffeine Intake and Stress Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Weekend_Screen_Time_Hours', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Weekend Screen Time Impact on Sleep Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Matrix of Health and Usage Metrics')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x='Daily_Phone_Hours', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Regression Analysis: Phone Usage vs Stress Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x='Sleep_Hours', y='Work_Productivity_Score', data=df)
plt.title(f'{plot_no}. Effect of Sleep Duration on Work Productivity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Gender', y='App_Usage_Count', data=df, estimator=np.mean)
plt.title(f'{plot_no}. Average App Usage Count by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Daily_Phone_Hours', data=df, hue='Occupation')
plt.title(f'{plot_no}. Age vs Daily Phone Usage Across Occupations')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Gender', y='Sleep_Hours', data=df)
plt.title(f'{plot_no}. Sleep Hours Distribution by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Social_Media_Hours', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Social Media Usage vs Stress Level Analysis')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Occupation', y='Stress_Level', data=df)
plt.title(f'{plot_no}. Stress Levels Across Different Occupations')
plt.xticks(rotation=45)
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='App_Usage_Count', y='Work_Productivity_Score', data=df)
plt.title(f'{plot_no}. App Usage Count vs Work Productivity Score')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Daily_Phone_Hours', y='Caffeine_Intake_Cups', data=df, hue='Stress_Level')
plt.title(f'{plot_no}. Phone Usage and Caffeine Intake Pattern by Stress Level')
show_fig()
plot_no += 1
```
