# Model Training

```python
# Calculate mean values of important features grouped by order_status
mean_comparison = df.groupby('order_status')[[
    'order_value',
    'delivery_time_minutes',
    'delivery_distance_km',
    'delivery_partner_rating'
]].mean()

print("Mean Feature Values by Order Status:\n")
mean_comparison
```

From the above output, we can clearly observe that the mean values of order_value, delivery_time_minutes, delivery_distance_km, and delivery_partner_rating are very similar across all three classes (Cancelled, Delayed, and Delivered).

The differences between these averages are extremely small and do not show any strong separation between the categories.

This indicates that the dataset does not contain strong predictive signals linking the input features to the target variable (order_status).

Because of this heavy similarity in feature distribution, machine learning models struggle to differentiate between classes, resulting in accuracy close to random guessing (~33% for three balanced classes).

This explains why improving the model architecture does not significantly improve performance.
