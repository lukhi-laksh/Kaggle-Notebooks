# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='selected', data=df)
plt.title(f'{plot_no}. Placement Selection Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='CGPA', hue='selected', kde=True)
plt.title(f'{plot_no}. CGPA Distribution by Placement Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='CGPA', data=df)
plt.title(f'{plot_no}. CGPA Comparison Between Selected and Not Selected Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='coding_test_score', data=df)
plt.title(f'{plot_no}. Coding Test Score Impact on Placement Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='interview_score', data=df)
plt.title(f'{plot_no}. Interview Score Difference by Placement Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='coding_test_score', y='interview_score', hue='selected', data=df)
plt.title(f'{plot_no}. Coding Test Score Versus Interview Score by Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='CGPA', y='resume_score', hue='selected', data=df)
plt.title(f'{plot_no}. CGPA Versus Resume Score in Placement Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='skills_score', y='projects_count', hue='selected', data=df)
plt.title(f'{plot_no}. Skills Score Versus Project Count by Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='internships_done', hue='selected', data=df)
plt.title(f'{plot_no}. Internship Experience and Placement Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='projects_count', hue='selected', data=df)
plt.title(f'{plot_no}. Project Count Distribution by Selection Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='placement_training', y='interview_score', data=df)
plt.title(f'{plot_no}. Placement Training Impact on Interview Score')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='placement_training', hue='selected', data=df)
plt.title(f'{plot_no}. Placement Training and Final Selection Outcome')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='communication_score', data=df)
plt.title(f'{plot_no}. Communication Score Difference by Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='aptitude_score', data=df)
plt.title(f'{plot_no}. Aptitude Score Comparison by Placement Result')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='linkedin_activity_score', y='github_score', hue='selected', data=df)
plt.title(f'{plot_no}. LinkedIn Activity Versus GitHub Score by Selection')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='selected', y='github_score', data=df)
plt.title(f'{plot_no}. GitHub Score Strength Among Selected Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 9))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numeric Placement Factors')
show_fig()
plot_no += 1
```
