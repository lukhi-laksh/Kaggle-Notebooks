# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.sort_values('forecast_year'), x='forecast_year', y='predicted_anxiety_rate', marker='o')
plt.title(f'{plot_no}. Anxiety Rate Trend Across Forecast Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.sort_values('forecast_year'), x='forecast_year', y='predicted_attention_span_minutes', marker='o')
plt.title(f'{plot_no}. Attention Span Change Across Forecast Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(data=df.sort_values('forecast_year'), x='forecast_year', y='predicted_ai_addiction_index', marker='o')
plt.title(f'{plot_no}. AI Addiction Index Growth Across Forecast Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='platform', y='predicted_anxiety_rate', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Anxiety Rate Comparison Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='platform', y='predicted_ai_addiction_index', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average AI Addiction Index Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='country', y='predicted_sleep_decline_pct', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Country-Wise Average Sleep Decline Percentage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='country', y='mental_health_recovery_rate', estimator='mean')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Country-Wise Mental Health Recovery Rate Comparison')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='platform', y='predicted_anxiety_rate')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Anxiety Rate Distribution Across Platforms')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='predicted_anxiety_rate', kde=True, bins=15)
plt.title(f'{plot_no}. Distribution of Predicted Anxiety Rate')
show_fig()
plot_no += 1
```

```python
df_platform_year = df.pivot_table(index='platform', columns='forecast_year', values='predicted_anxiety_rate', aggfunc='mean')
fig = plt.figure(figsize=(10,6))
sns.heatmap(df_platform_year, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Platform and Forecast Year Anxiety Risk Heatmap')
show_fig()
plot_no += 1
```

```python
df_country_platform = df.pivot_table(index='country', columns='platform', values='predicted_ai_addiction_index', aggfunc='mean')
fig = plt.figure(figsize=(10,6))
sns.heatmap(df_country_platform, annot=True, cmap='viridis')
plt.title(f'{plot_no}. Country and Platform AI Addiction Risk Heatmap')
show_fig()
plot_no += 1
```

```python
corr = df.select_dtypes(include='number').corr()
fig = plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Digital Mental Health Forecast Indicators')
show_fig()
plot_no += 1
```
