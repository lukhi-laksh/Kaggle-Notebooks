# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Monthly_Salary'], kde=True)
plt.title(f'{plot_no}. Distribution of Monthly Salary Across Employees')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='City', y='Monthly_Salary', data=df, estimator='mean')
plt.title(f'{plot_no}. Average Monthly Salary by City')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='City', y='Monthly_Salary', data=df, estimator='mean')
plt.title(f'{plot_no}. Average Monthly Salary by City')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Experience_Years', y='Monthly_Salary', hue='Department', data=df)
plt.title(f'{plot_no}. Relationship Between Experience and Salary by Department')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Monthly_Salary', hue='Gender', data=df)
plt.title(f'{plot_no}. Salary vs Age with Gender-Based Trend')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='Department', hue='Gender', data=df)
plt.title(f'{plot_no}. Gender Distribution Across Departments')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='Gender', y='Monthly_Salary', data=df)
plt.title(f'{plot_no}. Salary Density and Spread Across Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Education_Level', y='Experience_Years', data=df)
plt.title(f'{plot_no}. Experience Range Across Different Education Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='Education_Level', y='Monthly_Salary', data=df, estimator='mean')
plt.title(f'{plot_no}. Average Salary Comparison by Education Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df[['Experience_Years', 'Age', 'Monthly_Salary']].corr()
sns.heatmap(corr, annot=True)
plt.title(f'{plot_no}. Correlation Heatmap of Age, Experience, and Salary')
show_fig()
plot_no += 1
```
