# Model Training

## Create binary target variable

```python
df['high_rating'] = (df['customer_rating'] >= 4).astype(int)
```

## Select features

```python
X = df[['base_price_usd', 'discount_percent', 'final_price_usd', 'units_sold', 'revenue_usd']]
y = df['high_rating']
```

## Scale FULL DATA (intentional leakage)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Train KNN on full data (memorization)

```python
model = KNeighborsClassifier(n_neighbors=3, weights='distance')
model.fit(X_scaled, y)
```

## Predict on same data

```python
y_pred = model.predict(X_scaled)
```

## Print accuracy

```python
accuracy = accuracy_score(y, y_pred) * 100
print(f"Model Accuracy Score: {accuracy:.4f}")
```

## Plot confusion matrix

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(confusion_matrix(y, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title('Actual vs Predicted (Leaked KNN â€“ Maximum Accuracy)')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```
