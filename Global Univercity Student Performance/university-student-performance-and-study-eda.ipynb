# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['GPA'], kde=True)
plt.title(f'{plot_no}. Distribution of GPA Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['study_hours_per_day'], kde=True)
plt.title(f'{plot_no}. Daily Study Hours Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='study_hours_per_day', y='GPA', hue='country', data=df)
plt.title(f'{plot_no}. Relationship Between Study Hours and GPA')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='GPA', data=df)
plt.title(f'{plot_no}. GPA Comparison Across Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sleep_hours', y='GPA', data=df)
plt.title(f'{plot_no}. Impact of Sleep Hours on GPA')
show_fig()
plot_no += 1
```
