# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['calories'], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Calories Across Menu Items')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='brand_tier', y='price_usd_normalized', data=df)
plt.title(f'{plot_no}. Price Distribution Across Brand Tiers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='calories', y='total_fat_g', hue='item_category', data=df)
plt.title(f'{plot_no}. Relationship Between Calories and Total Fat')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='protein_g', y='calories', hue='brand_tier', data=df)
plt.title(f'{plot_no}. High Protein Foods Versus Calorie Density')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='city_tier', y='avg_rating', data=df)
plt.title(f'{plot_no}. Average Food Ratings by City Tier')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='item_category', data=df, order=df['item_category'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Most Common Food Categories in Dataset')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sugars_g', y='calories', hue='is_bestseller', data=df)
plt.title(f'{plot_no}. Sugar Levels Impact on Calorie Content and Bestseller Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='is_bestseller', y='avg_rating', data=df)
plt.title(f'{plot_no}. Rating Comparison Between Bestseller and Non-Bestseller Items')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(x='item_subcategory', y='sodium_mg', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Sodium Distribution Across Food Subcategories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='total_carbs_g', y='sugars_g', hue='item_category', data=df)
plt.title(f'{plot_no}. Carbohydrates Versus Sugar Content by Food Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='brand_name', y='calories', data=df, estimator=np.mean)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Calories Offered by Each Brand')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='price_usd_normalized', hue='brand_tier', fill=True)
plt.title(f'{plot_no}. Density Distribution of Food Prices by Brand Tier')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='serving_size_g', y='calories', size='total_fat_g', hue='item_category', data=df)
plt.title(f'{plot_no}. Serving Size Impact on Calories and Fat Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='country', y='sodium_mg', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Country-Wise Sodium Consumption Patterns')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(x='item_category', y='protein_g', data=df, estimator=np.mean)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Protein Content Across Food Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='price_usd_normalized', y='avg_rating', hue='brand_tier', data=df)
plt.title(f'{plot_no}. Relationship Between Product Pricing and Customer Ratings')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='item_category', y='sugars_g', data=df)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Sugar Content Comparison Across Food Categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.lineplot(x=df.groupby('city')['avg_rating'].mean().sort_values().index,
             y=df.groupby('city')['avg_rating'].mean().sort_values().values)
plt.xticks(rotation=90)
plt.title(f'{plot_no}. City-Wise Trend of Average Customer Ratings')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='sodium_mg', y='total_fat_g', hue='is_bestseller', data=df)
plt.title(f'{plot_no}. Sodium and Fat Combination in Bestseller Foods')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=np.number).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Nutritional and Business Features')
show_fig()
plot_no += 1
```
