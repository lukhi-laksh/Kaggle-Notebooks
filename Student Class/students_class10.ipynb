# Import Librarys

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Excel file

```python
df = pd.read_excel('students_class10_basic_dataset.xlsx')
df.head(5)
```

# Preprocessing

## Check Null Values

```python
df.isnull().sum()
```

## Check Shape

```python
print('DataFrame shape:', df.shape)
```

## Transforming DOB into Age

```python
df['DOB'] = pd.to_datetime(df['DOB'])
df['Age'] = (pd.Timestamp("today") - df['DOB']).dt.days // 365
```

## Check Duplicates

```python
df[df.duplicated()]
```

# EDA

## Univariate

### Which House has the most students?

```python
plt.figure(figsize=(6,4))
sns.countplot(x='House', data=df)
plt.show()
```

### 2. What is the gender distribution (Gender) in the dataset?

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Gender', data=df)
plt.show()
```

## Bivariate

### 1. Is there a correlation between Q_Maths and Q_Science scores?

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Q_Maths', y='Q_Science', data=df)
plt.show()
print("Correlation:", df['Q_Maths'].corr(df['Q_Science']))
```

### 2. How does Attendance_% vary between Male and Female students?

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Attendance_%', data=df)
plt.show()
```

### 3. Is there a relationship between Sports_Points_30 and Cultural_Points_20?

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Sports_Points_30', y='Cultural_Points_20', data=df)
plt.show()
print("Correlation:", df['Sports_Points_30'].corr(df['Cultural_Points_20']))
```

### 4. How do Q_History scores vary with Q_Civics scores?

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Q_History', y='Q_Civics', data=df)
plt.show()
print("Correlation:", df['Q_History'].corr(df['Q_Civics']))
```

### 5. Compare Science_Project_20 scores across different House categories.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='House', y='Science_Project_20', data=df)
plt.show()
```

### 6. Does higher Attendance_% lead to higher Q_Science scores?

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Attendance_%', y='Q_Science', data=df)
plt.show()
print("Correlation:", df['Attendance_%'].corr(df['Q_Science']))
```

### 7. Compare ArtCraft_20 scores between male and female students.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='ArtCraft_20', data=df)
plt.show()
```

### 8. Relationship between Discipline_10 and Volunteer_Hours.

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Discipline_10', y='Volunteer_Hours', data=df)
plt.show()
print("Correlation:", df['Discipline_10'].corr(df['Volunteer_Hours']))
```

### 9. Correlation between English_Activity_20 and Q_English scores.

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='English_Activity_20', y='Q_English', data=df)
plt.show()
print("Correlation:", df['English_Activity_20'].corr(df['Q_English']))
```

## Multivariate

### 1. How does Attendance_%, Gender, and House together affect Q_Maths scores?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='House', y='Q_Maths', hue='Gender', data=df)
plt.show()
```

### 2. Effect of Q_Science, Science_Project_20, and House on overall science performance.

```python
df['Science_Performance'] = df[['Q_Science','Science_Project_20']].mean(axis=1)
plt.figure(figsize=(8,5))
sns.boxplot(x='House', y='Science_Performance', data=df)
plt.show()
```

### 3. How do Q_English, English_Activity_20, and Gender together impact English proficiency?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Gender', y='Q_English', data=df)
sns.scatterplot(x='English_Activity_20', y='Q_English', hue='Gender', data=df)
plt.show()
```
