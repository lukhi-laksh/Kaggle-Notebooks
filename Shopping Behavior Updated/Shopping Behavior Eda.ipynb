# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title(f'{plot_no}. Age Distribution â€” concentration and outliers across customers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Purchase Amount (USD)', bins=40, kde=True)
plt.title(f'{plot_no}. Distribution of Purchase Amounts â€” skewness and common spend levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Category', y='Purchase Amount (USD)')
plt.title(f'{plot_no}. Purchase Amount by Category â€” variability and high-value categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Gender', y='Purchase Amount (USD)')
plt.title(f'{plot_no}. Gender vs Purchase Amount â€” comparison of spending behavior')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Category', y='Purchase Amount (USD)', estimator=np.mean, ci=None)
plt.title(f'{plot_no}. Average Purchase Amount per Category â€” identifying top spending categories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Subscription Status', y='Purchase Amount (USD)', estimator=np.mean, ci=None, order=df['Subscription Status'].value_counts().index)
plt.title(f'{plot_no}. Subscription Status vs Average Spend â€” subscription impact on value')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.title(f'{plot_no}. Category Frequency â€” most common product categories purchased')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Season', y='Purchase Amount (USD)', inner='quartile', order=df['Season'].value_counts().index)
plt.title(f'{plot_no}. Seasonal Variation in Purchase Amount â€” distribution by season')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.kdeplot(data=df, x='Purchase Amount (USD)', hue='Gender', fill=True)
plt.title(f'{plot_no}. Kernel Density of Purchase Amount by Gender â€” comparative spend density')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Shipping Type', y='Purchase Amount (USD)', order=df.groupby('Shipping Type')['Purchase Amount (USD)'].median().sort_values(ascending=False).index)
plt.title(f'{plot_no}. Shipping Type vs Purchase Amount â€” which shipping correlates with larger orders')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Payment Method', y='Review Rating', estimator=np.mean, ci=None, order=df.groupby('Payment Method')['Review Rating'].mean().sort_values(ascending=False).index)
plt.title(f'{plot_no}. Average Review Rating by Payment Method â€” satisfaction by payment channel')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
age_bins = pd.cut(df['Age'], bins=[0,18,25,35,45,55,65,100])
age_grouped = df.groupby(age_bins)['Purchase Amount (USD)'].mean().reset_index()
sns.barplot(data=age_grouped, x='Age', y='Purchase Amount (USD)') if False else plt.bar(age_grouped['Age'].astype(str), age_grouped['Purchase Amount (USD)'])
plt.title(f'{plot_no}. Average Spend by Age Group â€” which age brackets spend more')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=0.5, cmap='vlag')
plt.title(f'{plot_no}. Correlation Matrix of Numerical Features â€” relationships and multicollinearity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
cross = pd.crosstab(df['Promo Code Used'], df['Discount Applied'])
sns.heatmap(cross, annot=True, fmt='d')
plt.title(f'{plot_no}. Promo Code Usage vs Discount Applied â€” promo effectiveness and distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_locations = df['Location'].value_counts().nlargest(10).index
sns.barplot(data=df[df['Location'].isin(top_locations)].groupby('Location')['Purchase Amount (USD)'].mean().reset_index().sort_values('Purchase Amount (USD)', ascending=False), x='Location', y='Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Top 10 Locations by Average Purchase Amount â€” geographic spend hotspots')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Frequency of Purchases', order=df['Frequency of Purchases'].value_counts().index)
plt.title(f'{plot_no}. Purchase Frequency Distribution â€” how often customers buy')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Color', y='Purchase Amount (USD)', estimator=np.mean, ci=None, order=df.groupby('Color')['Purchase Amount (USD)'].mean().sort_values(ascending=False).index[:10])
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Top Colors by Average Purchase Amount â€” color influence on order value (top 10)')
show_fig()
plot_no += 1
```
