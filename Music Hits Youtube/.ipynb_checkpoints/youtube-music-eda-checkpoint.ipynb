# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
sns.barplot(x='channel', y='view_count', data=df.head(10))
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='duration', y='view_count', data=df.head(10))
plt.show()
```

```python
sns.barplot(x='categories', y='view_count', data=df.head(10))
plt.show()
```

```python
sns.boxplot(x='categories', y='duration', data=df.head(10))
plt.show()
```

```python
sns.barplot(x='channel', y='channel_follower_count', data=df.head(10))
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='channel_follower_count', y='view_count', hue='categories', data=df.head(10))
plt.show()
```

```python
sns.barplot(x='channel', y='duration', data=df.head(10))
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x='duration', y='channel_follower_count', size='view_count', data=df.head(10))
plt.show()
```

```python
sns.barplot(x='channel', y='duration', hue='categories', data=df.head(10))
plt.xticks(rotation=45)
plt.show()
```
