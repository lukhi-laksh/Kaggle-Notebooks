# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='job_title', order=df['job_title'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Distribution of Job Roles Across the Dataset')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='company_name', order=df['company_name'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Hiring Volume Comparison Across Companies')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['experience_required_years'], bins=15, kde=True)
plt.title(f'{plot_no}. Experience Requirement Distribution in Years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='job_title', y='salary_max_inr')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Maximum Salary Comparison by Job Role')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='employment_type', y='salary_min_inr')
plt.title(f'{plot_no}. Minimum Salary Variation by Employment Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='experience_required_years', y='salary_max_inr', hue='employment_type')
plt.title(f'{plot_no}. Relationship Between Experience and Maximum Salary')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='remote_option', y='salary_max_inr')
plt.title(f'{plot_no}. Salary Distribution Based on Remote Work Option')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
avg_salary = df.groupby('job_title')['salary_max_inr'].mean().sort_values(ascending=False)
sns.barplot(x=avg_salary.index, y=avg_salary.values)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Maximum Salary by Job Role')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
avg_exp = df.groupby('job_title')['experience_required_years'].mean().sort_values(ascending=False)
sns.barplot(x=avg_exp.index, y=avg_exp.values)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Experience Required per Job Role')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='employment_type', hue='remote_option')
plt.title(f'{plot_no}. Employment Type Distribution by Remote Option')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
pivot_table = df.pivot_table(values='salary_max_inr', index='company_name', columns='employment_type', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, fmt=".0f")
plt.title(f'{plot_no}. Heatmap of Average Maximum Salary by Company and Employment Type')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['salary_min_inr'], bins=20, kde=True)
plt.title(f'{plot_no}. Distribution of Minimum Salary Offered')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['salary_max_inr'], bins=20, kde=True)
plt.title(f'{plot_no}. Distribution of Maximum Salary Offered')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
salary_range = df['salary_max_inr'] - df['salary_min_inr']
sns.histplot(salary_range, bins=20, kde=True)
plt.title(f'{plot_no}. Salary Range Spread Across Job Listings')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='salary_min_inr', y='salary_max_inr', hue='job_title')
plt.title(f'{plot_no}. Correlation Between Minimum and Maximum Salary by Role')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='company_name', y='experience_required_years')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Experience Requirement Distribution Across Companies')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='remote_option', order=df['remote_option'].value_counts().index)
plt.title(f'{plot_no}. Remote Work Preference Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
company_salary = df.groupby('company_name')['salary_max_inr'].mean().sort_values(ascending=False)
sns.barplot(x=company_salary.index, y=company_salary.values)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Company-Wise Average Maximum Salary Comparison')
show_fig()
plot_no += 1
```
