# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title(f'{plot_no}. Age Distribution of Users Ordering Food')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['age'], kde=True)
plt.title(f'{plot_no}. Age Distribution of Users')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='city', data=df, order=df['city'].value_counts().index)
plt.title(f'{plot_no}. Order Volume by City')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='city', data=df)
plt.title(f'{plot_no}. Order Count Across Different Cities')
show_fig()
plot_no += 1
```
