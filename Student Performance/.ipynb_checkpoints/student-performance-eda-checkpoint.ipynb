# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x=df['Hours Studied'], y=df['Performance Index'], scatter_kws={'s':60}, line_kws={'lw':2})
plt.title(f'{plot_no}. Relationship between Hours Studied and Performance Index (linear trend & scatter)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x=df['Previous Scores'], y=df['Performance Index'], scatter_kws={'s':60}, line_kws={'lw':2})
plt.title(f'{plot_no}. How Previous Scores predict Performance Index (strength of association)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x=df['Sample Question Papers Practiced'], y=df['Performance Index'], scatter_kws={'s':80}, line_kws={'lw':2}, ci=None)
plt.title(f'{plot_no}. Impact of Number of Sample Question Papers Practiced on Performance Index')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(x=df['Sleep Hours'], y=df['Performance Index'], scatter_kws={'s':80}, line_kws={'lw':2}, lowess=True)
plt.title(f'{plot_no}. Non-linear association: Sleep Hours vs Performance Index (LOWESS fit)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Extracurricular Activities', y='Performance Index', data=df)
plt.title(f'{plot_no}. Distribution of Performance Index by Extracurricular Participation (boxplot)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Performance Index'], kde=True, bins=10)
plt.title(f'{plot_no}. Distribution of Performance Index with KDE (identify skewness and modes)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['Hours Studied'], kde=True, bins=8)
plt.title(f'{plot_no}. Distribution of Hours Studied across students (density + histogram)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x='Hours Studied', y='Previous Scores', data=df, s=80)
sns.kdeplot(x=df['Hours Studied'], y=df['Previous Scores'], levels=5, linewidths=1, fill=False)
plt.title(f'{plot_no}. Joint structure: Hours Studied vs Previous Scores with KDE contours (clusters & density)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced','Performance Index']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', square=True, cbar_kws={'shrink':0.8})
plt.title(f'{plot_no}. Correlation matrix between study, sleep, practice and Performance Index')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
model = sns.regplot(x=df['Hours Studied'], y=df['Performance Index'], scatter=False, line_kws={'lw':2})
residuals = df['Performance Index'] - (sns.regplot(x=df['Hours Studied'], y=df['Performance Index'], scatter=False).get_lines()[0].get_ydata() if False else df['Performance Index'] - df['Performance Index'])
import numpy as _np
coef = _np.polyfit(df['Hours Studied'], df['Performance Index'], 1)
pred = _np.polyval(coef, df['Hours Studied'])
resid = df['Performance Index'] - pred
plt.clf()
fig = plt.figure(figsize=(10,6))
sns.histplot(resid, kde=True, bins=8)
plt.title(f'{plot_no}. Residual distribution from linear fit: Hours Studied -> Performance Index (error diagnostics)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(x='Extracurricular Activities', y='Hours Studied', data=df)
plt.title(f'{plot_no}. Hours Studied distribution by Extracurricular Activities (do participants study more?)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
mean_perf_by_papers = df.groupby('Sample Question Papers Practiced')['Performance Index'].mean().reset_index()
sns.barplot(x='Sample Question Papers Practiced', y='Performance Index', data=mean_perf_by_papers)
plt.title(f'{plot_no}. Average Performance Index by Number of Sample Question Papers Practiced (comparative means)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
mean_perf_by_hours = df.groupby('Hours Studied')['Performance Index'].mean().reset_index().sort_values('Hours Studied')
sns.lineplot(x='Hours Studied', y='Performance Index', data=mean_perf_by_hours, marker='o')
plt.title(f'{plot_no}. Trend of Average Performance Index across Hours Studied (dose-response)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
mean_perf_by_sleep = df.groupby('Sleep Hours')['Performance Index'].mean().reset_index().sort_values('Sleep Hours')
sns.lineplot(x='Sleep Hours', y='Performance Index', data=mean_perf_by_sleep, marker='o')
plt.title(f'{plot_no}. Average Performance Index by Sleep Hours (optimal sleep signal)')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sizes = (df['Previous Scores'] - df['Previous Scores'].min() + 5) * 5
sns.scatterplot(x='Hours Studied', y='Previous Scores', size=sizes, sizes=(50,400), hue='Performance Index', data=df, palette='viridis', legend=False)
plt.title(f'{plot_no}. Hours Studied vs Previous Scores with point-size from Previous Scores and color by Performance Index')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
ax = sns.countplot(x='Extracurricular Activities', data=df)
ax2 = ax.twinx()
sns.pointplot(x='Extracurricular Activities', y='Performance Index', data=df, ax=ax2, order=df['Extracurricular Activities'].unique())
plt.title(f'{plot_no}. Count of Extracurricular Participation and Mean Performance Index overlay')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxenplot(x='Sample Question Papers Practiced', y='Performance Index', data=df)
plt.title(f'{plot_no}. Detailed distribution (boxen) of Performance Index by Sample Question Papers Practiced')
show_fig()
plot_no += 1
```
