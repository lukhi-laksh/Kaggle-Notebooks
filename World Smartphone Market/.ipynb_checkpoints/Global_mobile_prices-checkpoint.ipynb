# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
plt.figure(figsize=(10,6))
avg=df.groupby('brand')['price_usd'].mean().sort_values()
avg.plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Average Price (USD)')
plt.title(f'{plot_no}. Average Price per Brand')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(8,6))
df_box=df[['charging_watt','price_usd']].dropna()
df_box['charging_watt']=df_box['charging_watt'].astype(float)
sns.boxplot(x=pd.cut(df_box['charging_watt'], bins=4), y='price_usd', data=df_box)
plt.xlabel('Charging Watt (binned)')
plt.ylabel('Price (USD)')
plt.title(f'{plot_no}. Price Distribution by Charging Speed (binned)')
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(10,6))
order=df.groupby('brand')['price_usd'].median().sort_values().index
sns.boxplot(x='brand', y='price_usd', data=df, order=order)
plt.xlabel('Brand')
plt.ylabel('Price (USD)')
plt.title(f'{plot_no}. Price Distribution by Brand (ordered by median)')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(6,4))
counts=df['5g_support'].fillna('Unknown').value_counts()
counts.plot(kind='bar')
plt.xlabel('5G Support')
plt.ylabel('Count')
plt.title(f'{plot_no}. Count of 5G vs Non-5G Phones')
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(4,4))
df['os'].fillna('Unknown').value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title(f'{plot_no}. Operating System Market Share')
plt.ylabel('')
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(10,5))
proc=df['processor'].fillna('Unknown').value_counts().nlargest(10)
proc.plot(kind='bar')
plt.xlabel('Processor')
plt.ylabel('Count')
plt.title(f'{plot_no}. Top 10 Processors in Dataset')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(10,4))
order=['January','February','March','April','May','June','July','August','September','October','November','December']
months=df['release_month'].fillna('Unknown')
counts=months[months.isin(order)].value_counts().reindex(order).fillna(0)
counts.plot(kind='bar')
plt.xlabel('Release Month')
plt.ylabel('Count')
plt.title(f'{plot_no}. Phone Releases by Month')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(8,6))
num=df.select_dtypes(include=['number']).dropna(axis=1, how='all')
corr=num.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap (numeric features)')
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(10,6))
tab=pd.crosstab(df['brand'], df['5g_support'])
tab.plot(kind='bar', stacked=True)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title(f'{plot_no}. Stacked Count of 5G Support by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='brand', y='rating', data=df)
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.title(f'{plot_no}. Rating Distribution by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
show_fig()
plot_no += 1
```
