# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['sugar_g'], bins=10, kde=True)
plt.title(f'{plot_no}. Distribution of Sugar Content Across Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['calories'], bins=10, kde=True)
plt.title(f'{plot_no}. Calorie Distribution Highlighting Energy Density of Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='sugar_g', y='calories', s=120)
plt.title(f'{plot_no}. Relationship Between Sugar Content and Calories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.regplot(data=df, x='fiber_g_per_100g', y='sugar_g')
plt.title(f'{plot_no}. Fiber vs Sugar: Implications for Blood Sugar Control')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='taste_profile', y='sugar_g')
plt.title(f'{plot_no}. Sugar Distribution by Taste Profile')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='season', y='calories')
plt.title(f'{plot_no}. Seasonal Variation in Fruit Calories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='water_percent', y='calories', s=120)
plt.title(f'{plot_no}. Water Content vs Caloric Density')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='fiber_g_per_100g', y='calories', s=120)
plt.title(f'{plot_no}. Fiber Content vs Calories for Weight Management')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='fruit_name', y='fiber_g_per_100g')
plt.title(f'{plot_no}. Fiber Content Comparison Across Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='fruit_name', y='sugar_g')
plt.title(f'{plot_no}. Sugar Content Comparison Across Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='top_vitamin')
plt.title(f'{plot_no}. Distribution of Dominant Vitamins in Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='top_mineral')
plt.title(f'{plot_no}. Distribution of Key Minerals Across Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='avg_weight_g', y='calories', s=120)
plt.title(f'{plot_no}. Average Fruit Weight vs Total Calories')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.heatmap(df[['sugar_g','calories','acidity_pH','avg_weight_g','water_percent','fiber_g_per_100g','shelf_life_days']].corr(), annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Matrix of Nutritional and Physical Attributes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='shelf_life_days', y='water_percent', s=120)
plt.title(f'{plot_no}. Shelf Life vs Water Content Relationship')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='fruit_name', y='shelf_life_days')
plt.title(f'{plot_no}. Shelf Life Comparison and Storage Planning')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='acidity_pH', y='calories', s=120)
plt.title(f'{plot_no}. Impact of Acidity on Caloric Content')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='acidity_pH', y='sugar_g', s=120)
plt.title(f'{plot_no}. Acidity vs Sugar Balance in Fruits')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='season', y='sugar_g')
plt.title(f'{plot_no}. Seasonal Patterns in Sugar Content')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='avg_weight_g', y='sugar_g', s=120)
plt.title(f'{plot_no}. Fruit Weight vs Sugar Content per Serving')
show_fig()
plot_no += 1
```
