```python
import ollama

ollama.pull('mxbai-embed-large')

text = [
    "Machine learning algorithms use data to analyze and make predictions.",
    "Artificial intelligence is a broad field that enables computers to perform human-like tasks.",
    "Natural language processing focuses on the interaction between computers and human language.",
    "Computer vision allows machines to interpret and understand visual information from the world.",
    "Deep learning is a subset of machine learning using neural networks with many layers.",
    "Reinforcement learning involves an agent learning to make decisions by performing actions in an environment.",
    "Data science combines various fields to extract knowledge and insights from data."
]

response = ollama.embed(model='mxbai-embed-large', input=text)

embeddings = response['embeddings']

for i, emb in enumerate(embeddings):
    print(f"Generated vector embedding[{i}] of length: {len(emb)}")
    print(emb)
    print("\n")
```

```python
#Semantic Similarity Example
import ollama
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_embedding(text):
    response = ollama.embeddings(model='mxbai-embed-large', prompt=text)
    return response['embedding']

# Words to compare
words = ["computer", "mouse", "keyboard", "monitor"]
embeddings = [get_embedding(w) for w in words]

# Compare "computer" (index 0) with "mouse" (index 1)
sim_computer_mouse = cosine_similarity([embeddings[0]], [embeddings[1]])
# Compare "mouse" (index 1) with "keyboard" (index 2)
sim_mouse_keyboard = cosine_similarity([embeddings[1]], [embeddings[2]])

print(f"Similarity (computer vs mouse): {sim_computer_mouse[0][0]:.4f}")
print(f"Similarity (mouse vs keyboard): {sim_mouse_keyboard[0][0]:.4f}")
```

```python
import ollama
from sklearn.metrics.pairwise import cosine_similarity

def get_embed(text):
    return ollama.embeddings(model='mxbai-embed-large', prompt=text)['embedding']

# 1. Define our sentences
base_sentence = "The feline is sleeping."

# This uses different words but has the same meaning
similar_meaning = "A cat is taking a nap."

# This uses some similar words but has a different meaning
different_meaning = "The person is sleeping."

# 2. Generate Embeddings
vec_base = [get_embed(base_sentence)]
vec_sim  = [get_embed(similar_meaning)]
vec_diff = [get_embed(different_meaning)]

# 3. Compare
score_sim = cosine_similarity(vec_base, vec_sim)[0][0]
score_diff = cosine_similarity(vec_base, vec_diff)[0][0]

print(f"Similarity (Feline vs Cat Nap): {score_sim:.4f}")
print(f"Similarity (Feline vs Person): {score_diff:.4f}")
```

### New Example: Programming Language Similarities

Let's try comparing sentences related to programming languages to see how the embeddings capture their conceptual closeness.

```python
import ollama
from sklearn.metrics.pairwise import cosine_similarity

def get_embed(text):
    return ollama.embeddings(model='mxbai-embed-large', prompt=text)['embedding']

# Define new sentences for comparison
sentence_a = "Python is a versatile programming language used for web development and data science."
sentence_b = "Java is a powerful object-oriented language often used for enterprise applications."
sentence_c = "Data analysis with pandas in Python is very efficient."
sentence_d = "C++ is known for its performance and system-level programming capabilities."

# Generate embeddings for the new sentences
vec_a = [get_embed(sentence_a)]
vec_b = [get_embed(sentence_b)]
vec_c = [get_embed(sentence_c)]
vec_d = [get_embed(sentence_d)]

# Calculate similarities
sim_python_java = cosine_similarity(vec_a, vec_b)[0][0]
sim_python_data_analysis = cosine_similarity(vec_a, vec_c)[0][0]
sim_java_cpp = cosine_similarity(vec_b, vec_d)[0][0]
sim_data_analysis_java = cosine_similarity(vec_c, vec_b)[0][0]

print(f"Similarity (Python vs Java): {sim_python_java:.4f}")
print(f"Similarity (Python vs Data Analysis): {sim_python_data_analysis:.4f}")
print(f"Similarity (Java vs C++): {sim_java_cpp:.4f}")
print(f"Similarity (Data Analysis vs Java): {sim_data_analysis_java:.4f}")
```
