# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show() 
plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(np.log1p(df['duration_in_seconds']), bins=50, kde=True)
plt.title(f'{plot_no}. Log-transformed Call Duration Distribution  reveals multiplicative spread and outliers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
avg_day = df.groupby('day-of_week')['duration_in_seconds'].mean().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
sns.barplot(x=avg_day.index, y=avg_day.values)
plt.title(f'{plot_no}. Average Call Duration by Day of Week  highlights high-duration weekdays/weekends')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
count_day = df['day-of_week'].value_counts().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
sns.barplot(x=count_day.index, y=count_day.values)
plt.title(f'{plot_no}. Call Volume by Day of Week  busiest days by call count')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
avg_month = df.groupby('month')['duration_in_seconds'].mean().reindex(pd.date_range('2022-01-01', periods=12, freq='MS').strftime('%B'), fill_value=np.nan)
sns.barplot(x=avg_month.index, y=avg_month.values)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Call Duration by Month  seasonal patterns in call length')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
count_month = df['month'].value_counts().reindex(pd.date_range('2022-01-01', periods=12, freq='MS').strftime('%B'), fill_value=0)
sns.barplot(x=count_month.index, y=count_month.values)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Monthly Call Volume  months with highest activity')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_total = df.groupby('name')['duration_in_seconds'].sum().nlargest(10)
sns.barplot(y=top_total.index, x=top_total.values, orient='h')
plt.title(f'{plot_no}. Top 10 Callers by Total Call Duration  who accumulates most talk-time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top_count = df['name'].value_counts().nlargest(10)
sns.barplot(y=top_count.index, x=top_count.values, orient='h')
plt.title(f'{plot_no}. Top 10 Callers by Call Count  most frequently contacting numbers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sizes = df['category'].value_counts()
plt.pie(sizes.values, labels=sizes.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title(f'{plot_no}. Proportion of Calls by Category  categorical composition of calls')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
pivot = df.pivot_table(index='month', columns='category', values='dialled_phone_number', aggfunc='count').reindex(pd.date_range('2022-01-01', periods=12, freq='MS').strftime('%B')).fillna(0)
pivot.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title(f'{plot_no}. Stacked Monthly Call Counts by Category  monthly category composition')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
long_calls = df[df['duration_in_seconds'] > 600].groupby('category').size().sort_values(ascending=False)
sns.barplot(x=long_calls.index, y=long_calls.values)
plt.title(f'{plot_no}. Count of Long Calls (>10 minutes) by Category  which categories drive long conversations')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
short = df[df['duration_in_seconds'] < 60].groupby('category').size()
long = df[df['duration_in_seconds'] >= 60].groupby('category').size()
ratio = (short / (short + long)).fillna(0).reindex(df['category'].unique()).fillna(0)
sns.barplot(x=ratio.index, y=ratio.values)
plt.title(f'{plot_no}. Proportion of Short Calls (<60s) by Category  indicates quick contacts vs conversations')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(x='year', hue='category', data=df)
plt.title(f'{plot_no}. Yearly Call Counts by Category  changes in category composition over years')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
mean_by_name = df.groupby('name')['duration_in_seconds'].mean().nlargest(10)
sns.barplot(x=mean_by_name.values, y=mean_by_name.index, orient='h')
for i, v in enumerate(mean_by_name.values):
    plt.text(v, i, f'{v:.0f}', va='center')
plt.title(f'{plot_no}. Top 10 Names by Mean Call Duration  callers with longest average calls')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
top15 = df.groupby('name')['duration_in_seconds'].sum().nlargest(15)
sns.barplot(y=top15.index, x=top15.values, orient='h')
plt.title(f'{plot_no}. Horizontal Ranking: Top 15 Callers by Total Duration  ranked talk-time contributors')
show_fig()
plot_no += 1
```
