# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Map", order=df["Map"].value_counts().index)
plt.title(f'{plot_no}. Match Distribution Across Different Maps')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Tier", order=df["Tier"].value_counts().index)
plt.title(f'{plot_no}. Player Tier Distribution Showing Competitive Skill Spread')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df["Kills"], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Player Kills per Match')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df["Damage_Dealt"], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Total Damage Dealt by Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Kills", y="Damage_Dealt", hue="Win")
plt.title(f'{plot_no}. Relationship Between Kills and Damage with Win Indicator')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Win", y="Survival_Time_Minutes")
plt.title(f'{plot_no}. Survival Time Comparison Between Winning and Losing Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Distance_Travelled_km", y="Survival_Time_Minutes")
plt.title(f'{plot_no}. Impact of Travel Distance on Survival Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.violinplot(data=df, x="Tier", y="Kills")
plt.title(f'{plot_no}. Kill Distribution Across Competitive Player Tiers')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Headshot_Percentage", y="Kills")
plt.title(f'{plot_no}. Relationship Between Headshot Accuracy and Total Kills')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Assists", y="Kills")
plt.title(f'{plot_no}. Contribution of Assists to Total Kill Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Revives", y="Survival_Time_Minutes")
plt.title(f'{plot_no}. Influence of Team Revives on Player Survival Duration')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df["Longest_Kill_meters"], bins=30, kde=True)
plt.title(f'{plot_no}. Distribution of Longest Kill Distance Achieved by Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Heals_Used", y="Survival_Time_Minutes")
plt.title(f'{plot_no}. Relationship Between Healing Usage and Survival Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Map", y="Kills")
plt.title(f'{plot_no}. Kill Performance Variation Across Different Maps')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Map", y="Damage_Dealt")
plt.title(f'{plot_no}. Damage Output Comparison Across Maps')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Favorite_Weapon", order=df["Favorite_Weapon"].value_counts().index)
plt.title(f'{plot_no}. Most Preferred Weapons Among Players')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Win", y="Match_Rank")
plt.title(f'{plot_no}. Match Rank Distribution for Winners vs Non-Winners')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="Distance_Travelled_km", y="Kills")
plt.title(f'{plot_no}. Relationship Between Map Movement and Kill Performance')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(14, 7))
corr = df.select_dtypes(include="number").corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title(f'{plot_no}. Correlation Matrix of Gameplay Performance Metrics')
show_fig()
plot_no += 1
```
