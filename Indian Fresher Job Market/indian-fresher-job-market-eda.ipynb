# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='hiring_stage', order=df['hiring_stage'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Hiring Stage Distribution Shows Candidate Funnel Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='gender', hue='hiring_stage')
plt.title(f'{plot_no}. Hiring Stage Comparison Across Gender Groups')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title(f'{plot_no}. Candidate Age Distribution Highlights Applicant Experience Range')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='hiring_stage', y='age')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Age Variation Across Hiring Stages')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(data=df, x='cgpa', bins=20, kde=True)
plt.title(f'{plot_no}. CGPA Distribution Shows Academic Strength of Candidates')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='hiring_stage', y='cgpa')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. CGPA Comparison Across Hiring Outcomes')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='backlogs', hue='hiring_stage')
plt.title(f'{plot_no}. Backlogs Impact on Candidate Hiring Stage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='college', order=df['college'].value_counts().head(10).index)
plt.title(f'{plot_no}. Top Colleges Contributing Maximum Candidates')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='degree', order=df['degree'].value_counts().index)
plt.title(f'{plot_no}. Degree-Wise Candidate Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='branch', order=df['branch'].value_counts().head(10).index)
plt.title(f'{plot_no}. Branch-Wise Candidate Demand Pattern')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='sector', order=df['sector'].value_counts().index)
plt.title(f'{plot_no}. Sector-Wise Hiring Application Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='job_role', order=df['job_role'].value_counts().head(10).index)
plt.title(f'{plot_no}. Most Applied Job Roles Among Candidates')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='work_type', hue='hiring_stage')
plt.title(f'{plot_no}. Work Type Preference Compared With Hiring Stage')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, y='job_location', order=df['job_location'].value_counts().head(10).index)
plt.title(f'{plot_no}. Top Job Locations by Candidate Applications')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='hiring_stage', y='response_time_days')
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Response Time Difference Across Hiring Stages')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='referral_applied', y='response_time_days')
plt.title(f'{plot_no}. Referral Impact on Company Response Time')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
df['application_date'] = pd.to_datetime(df['application_date'])
df.groupby(df['application_date'].dt.year).size().plot(kind='line', marker='o')
plt.title(f'{plot_no}. Year-Wise Application Trend Shows Hiring Activity Growth')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap Shows Numeric Feature Relationships')
show_fig()
plot_no += 1
```
