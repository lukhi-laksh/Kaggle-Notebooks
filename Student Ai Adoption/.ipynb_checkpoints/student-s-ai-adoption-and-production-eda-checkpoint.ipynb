# EDA

```python
def show_fig():
    plt.tight_layout()
    plt.show()

plot_no = 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['cgpa'], kde=True, bins=25)
plt.title(f'{plot_no}. CGPA Distribution Across Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['productivity_score'], kde=True, bins=25)
plt.title(f'{plot_no}. Productivity Score Distribution Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.histplot(df['mental_wellbeing_score'], kde=True, bins=25)
plt.title(f'{plot_no}. Mental Wellbeing Score Distribution')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='daily_screen_time_hours', y='stress_level', hue='gender')
plt.title(f'{plot_no}. Screen Time vs Stress Level by Gender')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='ai_dependency_level', y='productivity_score')
plt.title(f'{plot_no}. Productivity Score Across AI Dependency Levels')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='placement_status', y='cgpa')
plt.title(f'{plot_no}. CGPA Comparison Across Placement Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='placement_status', y='expected_salary_lpa')
plt.title(f'{plot_no}. Expected Salary Variation by Placement Status')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='branch', y='cgpa', estimator='mean', errorbar=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average CGPA by Academic Branch')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='branch', y='daily_coding_hours', estimator='mean', errorbar=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Average Daily Coding Hours by Branch')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='preferred_programming_language', y='daily_coding_hours', estimator='mean', errorbar=None)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Coding Hours by Preferred Programming Language')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='favorite_ai_tool', order=df['favorite_ai_tool'].value_counts().index)
plt.xticks(rotation=45)
plt.title(f'{plot_no}. Most Preferred AI Tools Among Students')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.countplot(data=df, x='placement_status', hue='ai_dependency_level')
plt.title(f'{plot_no}. Placement Status Distribution by AI Dependency Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.barplot(data=df, x='internet_quality', y='daily_chatgpt_usage_hours', estimator='mean', errorbar=None)
plt.title(f'{plot_no}. Average ChatGPT Usage by Internet Quality')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='ai_replaceability_fear', y='stress_level', hue='ai_dependency_level')
plt.title(f'{plot_no}. AI Replaceability Fear vs Stress Level')
show_fig()
plot_no += 1
```

```python
fig = plt.figure(figsize=(15, 9))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title(f'{plot_no}. Correlation Heatmap of Numerical Student Performance Factors')
show_fig()
plot_no += 1
```
